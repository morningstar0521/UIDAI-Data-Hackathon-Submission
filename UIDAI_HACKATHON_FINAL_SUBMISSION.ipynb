````python
print("╔" + "═"*78 + "╗")
print("║" + " "*20 + "DATA INTEGRITY FRAMEWORK" + " "*28 + "║")
print("╚" + "═"*78 + "╝")

CANONICAL_STATES = {
    'Andaman and Nicobar Islands', 'Andhra Pradesh', 'Arunachal Pradesh', 'Assam',
    'Bihar', 'Chandigarh', 'Chhattisgarh', 'Dadra and Nagar Haveli and Daman and Diu',
    'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Jharkhand',
    'Karnataka', 'Kerala', 'Ladakh', 'Lakshadweep', 'Madhya Pradesh', 'Maharashtra',
    'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'NCT of Delhi', 'Odisha',
    'Puducherry', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telangana',
    'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal'
}

state_mapping = {
    'Balanagar': 'Telangana', 'BALANAGAR': 'Telangana', 'balanagar': 'Telangana',
    'Darbhanga': 'Bihar', 'DARBHANGA': 'Bihar', 'darbhanga': 'Bihar',
    'Jaipur': 'Rajasthan', 'JAIPUR': 'Rajasthan', 'jaipur': 'Rajasthan',
    'Madanapalle': 'Andhra Pradesh', 'MADANAPALLE': 'Andhra Pradesh',
    'Nagpur': 'Maharashtra', 'NAGPUR': 'Maharashtra', 'nagpur': 'Maharashtra',
    'Puttenahalli': 'Karnataka', 'PUTTENAHALLI': 'Karnataka',
    'Raja Annamalai Puram': 'Tamil Nadu', 'RAJA ANNAMALAI PURAM': 'Tamil Nadu',

    'Orissa': 'Odisha', 'ODISHA': 'Odisha', 'odisha': 'Odisha',

    'WEST BENGAL': 'West Bengal', 'WESTBENGAL': 'West Bengal', 'West  Bengal': 'West Bengal',
    'West Bangal': 'West Bengal', 'West Bengli': 'West Bengal', 'Westbengal': 'West Bengal',
    'West bengal': 'West Bengal', 'west bengal': 'West Bengal', 'west Bengal': 'West Bengal',

    'Jammu & Kashmir': 'Jammu and Kashmir', 'Jammu And Kashmir': 'Jammu and Kashmir',
    'J&K': 'Jammu and Kashmir', 'JAMMU AND KASHMIR': 'Jammu and Kashmir',

    'Andaman & Nicobar Islands': 'Andaman and Nicobar Islands',
    'Andaman and Nicobar': 'Andaman and Nicobar Islands',
    'A&N Islands': 'Andaman and Nicobar Islands',

    'Chhatisgarh': 'Chhattisgarh', 'CHHATTISGARH': 'Chhattisgarh',

    'Pondicherry': 'Puducherry', 'PUDUCHERRY': 'Puducherry',

    'Tamilnadu': 'Tamil Nadu', 'TAMIL NADU': 'Tamil Nadu',

    'Uttaranchal': 'Uttarakhand', 'UTTARAKHAND': 'Uttarakhand',

    'Dadra & Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'Dadra and Nagar Haveli': 'Dadra and Nagar Haveli and Daman and Diu',
    'Daman & Diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'Daman and Diu': 'Dadra and Nagar Haveli and Daman and Diu',
    'The Dadra And Nagar Haveli And Daman And Diu': 'Dadra and Nagar Haveli and Daman and Diu',

    'Delhi': 'NCT of Delhi', 'DELHI': 'NCT of Delhi', 'New Delhi': 'NCT of Delhi',

    'TELANGANA': 'Telangana', 'MAHARASHTRA': 'Maharashtra', 'KARNATAKA': 'Karnataka',
    'GUJARAT': 'Gujarat', 'RAJASTHAN': 'Rajasthan', 'KERALA': 'Kerala',
    'ANDHRA PRADESH': 'Andhra Pradesh', 'andhra pradesh': 'Andhra Pradesh',
    'UTTAR PRADESH': 'Uttar Pradesh', 'MADHYA PRADESH': 'Madhya Pradesh',
    'BIHAR': 'Bihar', 'JHARKHAND': 'Jharkhand', 'ASSAM': 'Assam',
    'PUNJAB': 'Punjab', 'HARYANA': 'Haryana',
}

INVALID_STATE_VALUES = {'100000', '100000 district', '', '?'}

district_mapping = {
    'Ananthapur': 'Anantapur',
    'Ananthapuramu': 'Anantapur',
    'Cuddapah': 'Y.S.R.',
    'Y. S. R': 'Y.S.R.',
    'Visakhapatanam': 'Visakhapatnam',
    'Spsr Nellore': 'Sri Potti Sriramulu Nellore',
    'Nellore': 'Sri Potti Sriramulu Nellore',
    'Mahabub Nagar': 'Mahabubnagar',
    'Mahbubnagar': 'Mahabubnagar',
    'Karim Nagar': 'Karimnagar',

    'Aurangabad(BH)': 'Aurangabad',
    'Aurangabad(bh)': 'Aurangabad',
    'Monghyr': 'Munger',
    'Purnea': 'Purnia',
    'Samstipur': 'Samastipur',
    'Sheikpura': 'Sheikhpura',
    'East Champaran': 'Purbi Champaran',
    'Pashchim Champaran': 'West Champaran',
    'Bhabua': 'Kaimur',

    'Lahaul and Spiti': 'Lahul and Spiti',
    'Lahul  Spiti': 'Lahul and Spiti',

    'Hazaribag': 'Hazaribagh',
    'Purbi Singhbhum': 'East Singhbhum',
    'Pashchimi Singhbhum': 'West Singhbhum',

    'Belgaum': 'Belagavi',
    'Bellary': 'Ballari',
    'Gulbarga': 'Kalaburagi',
    'Mysore': 'Mysuru',
    'Shimoga': 'Shivamogga',
    'Tumkur': 'Tumakuru',
    'Hasan': 'Hassan',
    'Chamarajanagar': 'Chamrajnagar',
    'Bangalore': 'Bengaluru Urban',
    'Bengaluru': 'Bengaluru Urban',

    'Hoshangabad': 'Narmadapuram',
    'Narsimhapur': 'Narsinghpur',
    'East Nimar': 'Khandwa',
    'West Nimar': 'Khargone',

    'Ahmadnagar': 'Ahilyanagar',
    'Ahmednagar': 'Ahilyanagar',
    'Ahmed Nagar': 'Ahilyanagar',
    'Aurangabad': 'Chhatrapati Sambhajinagar',
    'Osmanabad': 'Dharashiv',
    'Bid': 'Beed',

    'Anugul': 'Angul',
    'Anugal': 'Angul',
    'Nabarangapur': 'Nabarangpur',
    'Sonapur': 'Subarnapur',

    'Jalor': 'Jalore',
    'Jhunjhunun': 'Jhunjhunu',
    'Chittaurgarh': 'Chittorgarh',

    'Thoothukkudi': 'Thoothukkudi',
    'Tuticorin': 'Thoothukkudi',
    'Thiruvallur': 'Tiruvallur',

    'Jangaon': 'Jangoan',
    'Ranga Reddy': 'Rangareddy',
    'Medchal?malkajgiri': 'Medchal-Malkajgiri',
    'Medchalâmalkajgiri': 'Medchal-Malkajgiri',
    'Medchal−malkajgiri': 'Medchal-Malkajgiri',

    'Allahabad': 'Prayagraj',
    'Faizabad': 'Ayodhya',
    'Bara Banki': 'Barabanki',
    'Shravasti': 'Shrawasti',

    'Burdwan': 'Bardhaman',
    'Hooghiy': 'Hooghly',
    'Hugli': 'Hooghly',
    'Medinipur West': 'Paschim Medinipur',
    'West Midnapore': 'Paschim Medinipur',
    'Puruliya': 'Purulia',
    'South Twenty Four Parganas': 'South 24 Parganas',
    'South 24 Pargana': 'South 24 Parganas',
    'Dinajpur Dakshin': 'Dakshin Dinajpur',
}

INVALID_DISTRICT_VALUES = {'100000', '100000 District', '', '?', 'None', 'Na', 'N/A',
                           '5Th Cross', 'Near University Thana', 'Near Meera Hospital',
                           'Idpl Colony', 'Near Dhyana Ashram', 'Near Uday Nagar Nit Garden'}

region_mapping = {
    'North': ['NCT of Delhi', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Punjab',
              'Rajasthan', 'Uttarakhand', 'Uttar Pradesh', 'Chandigarh', 'Ladakh'],
    'South': ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Telangana', 'Puducherry',
              'Lakshadweep', 'Andaman and Nicobar Islands'],
    'East': ['Bihar', 'Jharkhand', 'Odisha', 'West Bengal'],
    'West': ['Goa', 'Gujarat', 'Maharashtra', 'Dadra and Nagar Haveli and Daman and Diu'],
    'Central': ['Chhattisgarh', 'Madhya Pradesh'],
    'Northeast': ['Arunachal Pradesh', 'Assam', 'Manipur', 'Meghalaya', 'Mizoram',
                  'Nagaland', 'Sikkim', 'Tripura']
}

def get_region(state):
    for region, states in region_mapping.items():
        if state in states:
            return region
    return 'Other'


def normalize_state(state_name):
    if state_name is None or pd.isna(state_name):
        return None

    state_str = str(state_name).strip()

    if state_str.lower() in {v.lower() for v in INVALID_STATE_VALUES} or state_str.isdigit():
        return None

    if state_str in state_mapping:
        return state_mapping[state_str]

    title_case = state_str.title()
    if title_case in state_mapping:
        return state_mapping[title_case]

    if title_case in CANONICAL_STATES:
        return title_case

    cleaned = ' '.join(state_str.split()).title()
    cleaned = cleaned.replace(' And ', ' and ')
    if cleaned in CANONICAL_STATES:
        return cleaned

    return None

def normalize_district(district_name):
    if district_name is None or pd.isna(district_name):
        return None
    district_str = str(district_name).strip()

    district_str = district_str.replace('*', '').replace('#', '').replace('@', '').strip()

    if district_str.lower() in {v.lower() for v in INVALID_DISTRICT_VALUES} or district_str.isdigit():
        return None

    district_title = ' '.join(district_str.split()).title()

    if district_title in district_mapping:
        return district_mapping[district_title]

    if district_str in district_mapping:
        return district_mapping[district_str]

    for key, value in district_mapping.items():
        if key.lower() == district_str.lower():
            return value

    return district_title

def clean_dataset(df, name):

    if df is None or len(df) == 0:
        print(f"   {name}: Empty dataset")
        return pd.DataFrame()

    df_clean = df.copy()
    initial_count = len(df_clean)

    df_clean['date'] = pd.to_datetime(df_clean['date'], format='mixed', errors='coerce')
    invalid_dates = df_clean['date'].isna().sum()
    df_clean = df_clean.dropna(subset=['date'])

    df_clean['state'] = df_clean['state'].apply(normalize_state)
    invalid_states = df_clean['state'].isna().sum()
    df_clean = df_clean.dropna(subset=['state'])

    df_clean['district'] = df_clean['district'].apply(normalize_district)
    invalid_districts = df_clean['district'].isna().sum()
    df_clean = df_clean.dropna(subset=['district'])

    before_dedup = len(df_clean)
    df_clean = df_clean.drop_duplicates()
    duplicates_removed = before_dedup - len(df_clean)

    numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
    df_clean[numeric_cols] = df_clean[numeric_cols].fillna(0)

    df_clean['region'] = df_clean['state'].apply(get_region)
    df_clean['year'] = df_clean['date'].dt.year
    df_clean['month'] = df_clean['date'].dt.month
    df_clean['month_name'] = df_clean['date'].dt.month_name()
    df_clean['day_of_week'] = df_clean['date'].dt.dayofweek
    df_clean['day_name'] = df_clean['date'].dt.day_name()
    df_clean['is_weekend'] = df_clean['day_of_week'].isin([5, 6]).astype(int)
    df_clean['quarter'] = df_clean['date'].dt.quarter
    df_clean['week_of_year'] = df_clean['date'].dt.isocalendar().week

    final_count = len(df_clean)

    print(f"""
    {name} Cleaning Report:
   ──────────────────────────────────────────
   - Initial records:      {initial_count:>12,}
   - Invalid dates:        {invalid_dates:>12,}
   - Invalid states:       {invalid_states:>12,}
   - Invalid districts:    {invalid_districts:>12,}
   - Duplicates removed:   {duplicates_removed:>12,}
   - Final clean records:  {final_count:>12,}
   - Data retention rate:  {final_count/initial_count*100:>11.1f}%
""")
    return df_clean

print("\n Applying Data Integrity Framework...")

df_enroll = clean_dataset(df_enroll_raw, 'ENROLLMENT')
df_demo = clean_dataset(df_demo_raw, 'DEMOGRAPHIC')
df_bio = clean_dataset(df_bio_raw, 'BIOMETRIC')

if len(df_enroll) > 0:
    enroll_cols = ['age_0_5', 'age_5_17', 'age_18_greater']
    for col in enroll_cols:
        if col not in df_enroll.columns:
            df_enroll[col] = 0
    df_enroll['total_enrollment'] = df_enroll['age_0_5'] + df_enroll['age_5_17'] + df_enroll['age_18_greater']
    df_enroll['child_enrollment'] = df_enroll['age_0_5'] + df_enroll['age_5_17']
    df_enroll['adult_enrollment'] = df_enroll['age_18_greater']

if len(df_demo) > 0:
    if 'demo_age_5_17' in df_demo.columns and 'demo_age_17_' in df_demo.columns:
        df_demo['demo_child'] = df_demo['demo_age_5_17']
        df_demo['demo_adult'] = df_demo['demo_age_17_']
        df_demo['total_demo'] = df_demo['demo_child'] + df_demo['demo_adult']
    else:
        df_demo['total_demo'] = 0

if len(df_bio) > 0:
    if 'bio_child' in df_bio.columns and 'bio_adult' in df_bio.columns:
        df_bio['total_bio'] = df_bio['bio_child'] + df_bio['bio_adult']
    else:
        df_bio['total_bio'] = 0

print(" Data cleaning complete - all datasets validated and standardized!")

print("\n" + "═"*70)
print(" VALIDATION SUMMARY:")
print("═"*70)

print("\n STATE VALIDATION:")
for df, name in [(df_enroll, 'Enrollment'), (df_demo, 'Demographic'), (df_bio, 'Biometric')]:
    if len(df) > 0:
        unique_states = set(df['state'].unique())
        invalid = unique_states - CANONICAL_STATES
        if invalid:
            print(f"    {name}: Non-canonical states found: {invalid}")
        else:
            print(f"    {name}: All {len(unique_states)} states are canonical")

print("\n DISTRICT VALIDATION:")
for df, name in [(df_enroll, 'Enrollment'), (df_demo, 'Demographic'), (df_bio, 'Biometric')]:
    if len(df) > 0:
        unique_districts = df[['state', 'district']].drop_duplicates()
        print(f"    {name}: {len(unique_districts)} unique (state, district) pairs")

print("\n\n" + "╔" + "═"*78 + "╗")
print("║" + " "*15 + " GEOGRAPHIC COVERAGE AFTER CLEANING" + " "*22 + "║")
print("╚" + "═"*78 + "╝")

enroll_states = set(df_enroll['state'].unique()) if len(df_enroll) > 0 else set()
demo_states = set(df_demo['state'].unique()) if len(df_demo) > 0 else set()
bio_states = set(df_bio['state'].unique()) if len(df_bio) > 0 else set()

all_states = sorted(enroll_states | demo_states | bio_states)

print(f"\n STATES & UNION TERRITORIES: {len(all_states)}")
print("─"*70)

for i, state in enumerate(all_states, 1):
    in_enroll = '✅' if state in enroll_states else '❌'
    in_demo = '✅' if state in demo_states else '❌'
    in_bio = '✅' if state in bio_states else '❌'
    print(f"   {i:>2}. {state:<40} [E:{in_enroll} D:{in_demo} B:{in_bio}]")

print(f"\n   Legend: E=Enrollment, D=Demographic, B=Biometric")

enroll_districts = set(df_enroll[['state', 'district']].dropna().apply(tuple, axis=1)) if len(df_enroll) > 0 else set()
demo_districts = set(df_demo[['state', 'district']].dropna().apply(tuple, axis=1)) if len(df_demo) > 0 else set()
bio_districts = set(df_bio[['state', 'district']].dropna().apply(tuple, axis=1)) if len(df_bio) > 0 else set()

all_districts = enroll_districts | demo_districts | bio_districts
full_coverage_districts = enroll_districts & demo_districts & bio_districts

print(f"\n\n DISTRICT SUMMARY:")
print("─"*70)
print(f"   Total unique (state, district) pairs: {len(all_districts)}")
print(f"   Districts with FULL coverage [E,D,B]: {len(full_coverage_districts)}")

district_counts = {}
for state, district in all_districts:
    if state and district:
        district_counts[state] = district_counts.get(state, 0) + 1

print(f"\n DISTRICTS PER STATE (Top 10):")
print("─"*70)
for state, count in sorted(district_counts.items(), key=lambda x: x[1], reverse=True)[:10]:
    bar = '█' * min(count // 3, 25)
    print(f"   {state:<35} {count:>4} {bar}")

print(f"\n   ... and {len(district_counts) - 10} more states")

print(f"""
════════════════════════════════════════════════════════════════════════════════
SECTION 2 COMPLETE
════════════════════════════════════════════════════════════════════════════════
   States/UTs:              {len(all_states):>5}
   Total Districts:         {len(all_districts):>5}
   Full Coverage Districts: {len(full_coverage_districts):>5}

   Enrollment Records:      {len(df_enroll):>10,}
   Demographic Records:     {len(df_demo):>10,}
   Biometric Records:       {len(df_bio):>10,}
════════════════════════════════════════════════════════════════════════════════
""")
````

##  SECTION 3: Unified Master Dataset & Cross-Dataset Linkage

### Why Unification Matters for Predictive Governance

Siloed datasets tell partial stories. A district with **high enrollment** but **low biometric compliance** signals operational failure. High **demographic updates** without corresponding **biometric verification** signals fraud risk. Only by unifying can we detect these **governance-critical patterns**.

### Linkage Strategy
- **Primary Keys**: State + District (normalized)
- **Join Type**: Outer join (preserve all entities, flag missing coverage)
- **Derived Metrics**: Cross-dataset ratios that surface anomalies

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*20 + " CREATING UNIFIED MASTER DATASET" + " "*21 + "║")
print("╚" + "═"*78 + "╝")

epsilon = 1e-6

if len(df_enroll) > 0:
    df_enroll_agg = df_enroll.groupby(['state', 'district']).agg({
        'age_0_5': 'sum', 'age_5_17': 'sum', 'age_18_greater': 'sum',
        'total_enrollment': 'sum', 'child_enrollment': 'sum', 'adult_enrollment': 'sum',
        'pincode': 'nunique', 'date': ['nunique', 'min', 'max']
    }).reset_index()
    df_enroll_agg.columns = ['state', 'district', 'enroll_0_5', 'enroll_5_17', 'enroll_adult',
                             'total_enrollment', 'child_enrollment', 'adult_enrollment_sum',
                             'enroll_pincodes', 'enroll_active_days', 'enroll_first_date', 'enroll_last_date']
    print(f"    Enrollment aggregated: {len(df_enroll_agg)} districts")
else:
    df_enroll_agg = pd.DataFrame(columns=['state', 'district'])
    print(f"    Enrollment data not available")

if len(df_demo) > 0:
    df_demo_agg = df_demo.groupby(['state', 'district']).agg({
        'demo_child': 'sum', 'demo_adult': 'sum', 'total_demo': 'sum',
        'pincode': 'nunique', 'date': ['nunique', 'min', 'max']
    }).reset_index()
    df_demo_agg.columns = ['state', 'district', 'demo_child', 'demo_adult', 'total_demo',
                           'demo_pincodes', 'demo_active_days', 'demo_first_date', 'demo_last_date']
    print(f"    Demographic aggregated: {len(df_demo_agg)} districts")
else:
    df_demo_agg = pd.DataFrame(columns=['state', 'district'])
    print(f"    Demographic data not available")

if len(df_bio) > 0:
    df_bio_agg = df_bio.groupby(['state', 'district']).agg({
        'bio_child': 'sum', 'bio_adult': 'sum', 'total_bio': 'sum',
        'mbu_compliance_index': 'mean',
        'pincode': 'nunique', 'date': ['nunique', 'min', 'max']
    }).reset_index()
    df_bio_agg.columns = ['state', 'district', 'bio_child', 'bio_adult', 'total_bio',
                          'avg_mbu_compliance', 'bio_pincodes', 'bio_active_days',
                          'bio_first_date', 'bio_last_date']
    print(f"    Biometric aggregated: {len(df_bio_agg)} districts")
else:
    df_bio_agg = pd.DataFrame(columns=['state', 'district'])
    print(f"    Biometric data not available")

print("\n Merging datasets on [state, district] keys...")

df_unified = df_enroll_agg.copy() if len(df_enroll_agg) > 0 else pd.DataFrame()

if len(df_demo_agg) > 0:
    if len(df_unified) > 0:
        df_unified = pd.merge(df_unified, df_demo_agg, on=['state', 'district'], how='outer')
    else:
        df_unified = df_demo_agg.copy()

if len(df_bio_agg) > 0:
    if len(df_unified) > 0:
        df_unified = pd.merge(df_unified, df_bio_agg, on=['state', 'district'], how='outer')
    else:
        df_unified = df_bio_agg.copy()

numeric_cols = df_unified.select_dtypes(include=[np.number]).columns
df_unified[numeric_cols] = df_unified[numeric_cols].fillna(0)

df_unified['region'] = df_unified['state'].apply(get_region)

df_unified['has_enrollment'] = df_unified['total_enrollment'] > 0
df_unified['has_demographic'] = df_unified['total_demo'] > 0
df_unified['has_biometric'] = df_unified['total_bio'] > 0
df_unified['data_completeness'] = df_unified['has_enrollment'].astype(int) + \
                                   df_unified['has_demographic'].astype(int) + \
                                   df_unified['has_biometric'].astype(int)

df_unified['total_activity'] = df_unified['total_enrollment'] + df_unified['total_demo'] + df_unified['total_bio']

print(f"\n{'═'*80}")
print(" UNIFIED MASTER DATASET CREATED")
print(f"{'═'*80}")
print(f"""
   Dimension                    Count
   ────────────────────────────────────
   Total Districts:            {len(df_unified):>6,}
   States/UTs:                 {df_unified['state'].nunique():>6}

   Data Completeness:
   • Districts in all 3 datasets:    {len(df_unified[df_unified['data_completeness']==3]):>4}
   • Districts in 2 datasets:        {len(df_unified[df_unified['data_completeness']==2]):>4}
   • Districts in 1 dataset:         {len(df_unified[df_unified['data_completeness']==1]):>4}

   Coverage Gap Alert:
   • Missing enrollment data:        {len(df_unified[~df_unified['has_enrollment']]):>4} districts
   • Missing demographic data:       {len(df_unified[~df_unified['has_demographic']]):>4} districts
   • Missing biometric data:         {len(df_unified[~df_unified['has_biometric']]):>4} districts
""")
````

Raw numbers are misleading. A state with 1M enrollments but 50M population has 2% coverage. A state with 100K enrollments and 500K population has 20% coverage. This section focuses on **relative metrics** that drive governance decisions.

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*15 + "EXPLORATORY ANALYSIS: CURRENT STATE" + " "*20 + "║")
print("╚" + "═"*78 + "╝")

print("\n STATE-LEVEL ACTIVITY DISTRIBUTION")
print("─"*70)

state_summary = df_unified.groupby('state').agg({
    'total_enrollment': 'sum',
    'total_demo': 'sum',
    'total_bio': 'sum',
    'total_activity': 'sum',
    'district': 'count'
}).reset_index()
state_summary.columns = ['State', 'Enrollments', 'Demographics', 'Biometrics', 'Total Activity', 'Districts']
state_summary = state_summary.sort_values('Total Activity', ascending=False)

print("\n TOP 10 STATES BY TOTAL AADHAAR ACTIVITY:")
print("─"*70)
top_10 = state_summary.head(10)
for i, row in top_10.iterrows():
    pct = row['Total Activity'] / state_summary['Total Activity'].sum() * 100
    print(f"   {row['State']:<25} | {row['Total Activity']:>12,.0f} ({pct:>5.1f}%) | {row['Districts']:>3} districts")

print("\n BOTTOM 5 STATES BY ACTIVITY (Potential Coverage Gaps):")
print("─"*70)
bottom_5 = state_summary.tail(5)
for i, row in bottom_5.iterrows():
    pct = row['Total Activity'] / state_summary['Total Activity'].sum() * 100
    print(f"   {row['State']:<25} | {row['Total Activity']:>12,.0f} ({pct:>5.2f}%) | {row['Districts']:>3} districts")

print("\n\n REGIONAL DISPARITY ANALYSIS")
print("─"*70)

regional_summary = df_unified.groupby('region').agg({
    'total_enrollment': 'sum',
    'total_demo': 'sum',
    'total_bio': 'sum',
    'total_activity': 'sum',
    'district': 'count',
    'state': 'nunique'
}).reset_index()
regional_summary.columns = ['Region', 'Enrollments', 'Demographics', 'Biometrics', 'Total Activity', 'Districts', 'States']
regional_summary['Activity_per_District'] = regional_summary['Total Activity'] / regional_summary['Districts']
regional_summary = regional_summary.sort_values('Activity_per_District', ascending=False)

print("\n REGIONAL ACTIVITY DENSITY (Activity per District):")
print("─"*70)
for i, row in regional_summary.iterrows():
    status = '🟢' if row['Activity_per_District'] > regional_summary['Activity_per_District'].median() else '🔴'
    print(f"   {status} {row['Region']:<12} | {row['Activity_per_District']:>12,.0f} per district | {row['Districts']:>3} districts | {row['States']:>2} states")

max_activity = regional_summary['Activity_per_District'].max()
min_activity = regional_summary['Activity_per_District'].max()
disparity_ratio = regional_summary['Activity_per_District'].max() / (regional_summary['Activity_per_District'].min() + epsilon)
print(f"\n REGIONAL DISPARITY RATIO: {disparity_ratio:.1f}x")
print(f"   (Highest activity region has {disparity_ratio:.1f}x more activity per district than lowest)")

print("\n\n AGE-GROUP ENROLLMENT ANALYSIS (MBU Pipeline)")
print("─"*70)

if 'enroll_0_5' in df_unified.columns:
    age_totals = {
        'Age 0-5 (Infants)': df_unified['enroll_0_5'].sum(),
        'Age 5-17 (Children)': df_unified['enroll_5_17'].sum(),
        'Age 18+ (Adults)': df_unified['enroll_adult'].sum()
    }
    total_enrollment = sum(age_totals.values())

    print("\n ENROLLMENT BY AGE GROUP:")
    print("─"*70)
    for age_group, count in age_totals.items():
        pct = count / total_enrollment * 100 if total_enrollment > 0 else 0
        bar = '█' * int(pct / 2)
        print(f"   {age_group:<20} | {count:>12,.0f} ({pct:>5.1f}%) {bar}")

    child_total = age_totals['Age 0-5 (Infants)'] + age_totals['Age 5-17 (Children)']
    print(f"\n MBU PIPELINE ALERT:")
    print(f"   • {child_total:,.0f} children enrolled today")
    print(f"   • These will require MANDATORY biometric update in 5-7 years")
    print(f"   • Represents {child_total/total_enrollment*100:.1f}% of all enrollments")

print("\n\n TEMPORAL PATTERN ANALYSIS")
print("─"*70)

if len(df_enroll) > 0 and 'month_name' in df_enroll.columns:
    monthly_pattern = df_enroll.groupby('month_name')['total_enrollment'].sum().reset_index()
    month_order = ['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December']
    monthly_pattern['month_name'] = pd.Categorical(monthly_pattern['month_name'], categories=month_order, ordered=True)
    monthly_pattern = monthly_pattern.sort_values('month_name')

    peak_month = monthly_pattern.loc[monthly_pattern['total_enrollment'].idxmax(), 'month_name']
    low_month = monthly_pattern.loc[monthly_pattern['total_enrollment'].idxmin(), 'month_name']

    print(f"\n ENROLLMENT SEASONALITY:")
    print(f"   • Peak Month: {peak_month}")
    print(f"   • Low Month: {low_month}")

    weekend_pct = df_enroll[df_enroll['is_weekend'] == 1]['total_enrollment'].sum() / df_enroll['total_enrollment'].sum() * 100
    print(f"\n WEEKEND ACCESS:")
    print(f"   • Weekend Enrollment: {weekend_pct:.1f}% of total")
    if weekend_pct < 10:
        print(f"   LOW weekend access - potential barrier for working population")
````

- "Which levers matter most?" - the question every policymaker asks

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*12 + "CORRELATION ANALYSIS & FEATURE IMPORTANCE" + " "*17 + "║")
print("╚" + "═"*78 + "╝")

import seaborn as sns
from sklearn.ensemble import RandomForestRegressor

print("""
    WHY THIS MATTERS FOR GOVERNANCE

   "If I can only fix ONE thing, what should it be?"

   This analysis answers:
   - Which factors are most correlated with risk?
   - What features best predict enrollment stress?
   - Where should limited resources be focused?
""")

print("\n CORRELATION MATRIX:")
print("═"*70)

numeric_cols = ['total_enrollment', 'total_demo', 'total_bio',
                'security_ratio', 'migration_intensity', 'AESI',
                'total_activity', 'data_completeness']

available_cols = [col for col in numeric_cols if col in df_unified.columns]
corr_data = df_unified[available_cols].dropna()

corr_matrix = corr_data.corr()

fig, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt='.2f',
    cmap='RdYlGn_r',
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={'label': 'Correlation Coefficient'},
    annot_kws={'size': 10}
)

plt.title(' Correlation Matrix: What Factors Drive Risk?\n(Red = Positive Correlation | Green = Negative)',
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("\n KEY CORRELATION INSIGHTS:")
print("─"*70)
if 'AESI' in corr_matrix.columns:
    aesi_corr = corr_matrix['AESI'].drop('AESI').sort_values(key=abs, ascending=False)

    print("\n   FACTORS MOST CORRELATED WITH AESI (Risk Score):")
    print("   ─"*50)
    for feature, corr_val in aesi_corr.head(5).items():
        direction = "↑ increases risk" if corr_val > 0 else "↓ reduces risk"
        strength = "Strong" if abs(corr_val) > 0.5 else "Moderate" if abs(corr_val) > 0.3 else "Weak"
        print(f"      • {feature}: {corr_val:.3f} ({strength} - {direction})")

print("\n\n FEATURE IMPORTANCE ANALYSIS (Random Forest):")
print("═"*70)
feature_cols = ['total_enrollment', 'total_demo', 'total_bio',
                'security_ratio', 'migration_intensity', 'data_completeness']
available_features = [col for col in feature_cols if col in df_unified.columns]

if 'AESI' in df_unified.columns and len(available_features) >= 3:
    X = df_unified[available_features].fillna(0)
    y = df_unified['AESI'].fillna(df_unified['AESI'].median())
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    rf_model.fit(X, y)
    importance_df = pd.DataFrame({
        'Feature': available_features,
        'Importance': rf_model.feature_importances_
    }).sort_values('Importance', ascending=True)
    fig, ax = plt.subplots(figsize=(10, 6))

    colors = ['#27ae60' if imp < 0.15 else '#f39c12' if imp < 0.25 else '#e74c3c'
              for imp in importance_df['Importance']]

    bars = ax.barh(importance_df['Feature'], importance_df['Importance'], color=colors)

    ax.set_xlabel('Importance Score', fontsize=12)
    ax.set_title('🎯 Feature Importance: What Predicts Enrollment Stress?\n(Higher = More Important for AESI Prediction)',
                 fontsize=14, fontweight='bold')
    for bar, val in zip(bars, importance_df['Importance']):
        ax.text(val + 0.01, bar.get_y() + bar.get_height()/2,
                f'{val:.3f}', va='center', fontsize=10)
    plt.tight_layout()
    plt.show()
    print("\n FEATURE IMPORTANCE INSIGHTS:")
    print("─"*70)

    top_features = importance_df.nlargest(3, 'Importance')
    print("\n    TOP 3 MOST IMPORTANT PREDICTORS OF RISK:")
    for i, (_, row) in enumerate(top_features.iterrows(), 1):
        print(f"      {i}. {row['Feature']}: {row['Importance']:.1%} importance")

    print(f"""

   GOVERNANCE IMPLICATION:
   ──────────────────────────────────────────────────────────────────────
   Focus intervention resources on the top predictors:
""")

    for _, row in top_features.iterrows():
        feature = row['Feature']
        if 'security' in feature.lower():
            print(f"      • {feature}: Enforce biometric compliance protocols")
        elif 'migration' in feature.lower():
            print(f"      • {feature}: Investigate address change anomalies")
        elif 'enrollment' in feature.lower():
            print(f"      • {feature}: Capacity planning for new registrations")
        elif 'demo' in feature.lower():
            print(f"      • {feature}: Monitor demographic update patterns")
        elif 'bio' in feature.lower():
            print(f"      • {feature}: Strengthen biometric verification")
        else:
            print(f"      • {feature}: Monitor and optimize")

else:
    print("   Insufficient features for importance analysis")

print("\n\n INTERACTIVE CORRELATION VIEW:")
print("═"*70)

fig = px.imshow(
    corr_matrix,
    text_auto='.2f',
    color_continuous_scale='RdYlGn_r',
    aspect='auto',
    title='📊 Interactive Correlation Matrix'
)

fig.update_layout(
    title={
        'text': ' Correlation Matrix: Factor Relationships<br><sup>Click to explore | Red = Positive | Green = Negative</sup>',
        'y': 0.95,
        'x': 0.5,
        'xanchor': 'center'
    },
    height=500
)

fig.show()

print(f"""
════════════════════════════════════════════════════════════════════════════════
SUMMARY: KEY DRIVERS OF AADHAAR ENROLLMENT STRESS
════════════════════════════════════════════════════════════════════════════════

   ACTIONABLE INSIGHT:
   "If you can only fix ONE thing, focus on the highest-importance feature."

   This analysis proves which levers matter most for governance intervention,
   enabling data-driven resource allocation.

════════════════════════════════════════════════════════════════════════════════
Correlation & Feature Importance analysis complete!
════════════════════════════════════════════════════════════════════════════════
""")
````

##  SECTION 5: DATA INTEGRITY & FRAUD PREVENTION MODULE

> **"Prevention is cheaper than prosecution."**

This module implements automated surveillance of enrollment patterns to identify potential data integrity issues and fraudulent activities. The system employs statistical anomaly detection to flag suspicious centers in real-time.

###  Detection Capabilities

| Detection Type | Method | Significance |
|----------------|--------|--------------|
| **Ghost Centers** | Isolation Forest | Flags centers with anomalous activity patterns |
| **Biometric Bypass** | Security Ratio < 0.3 | Identifies operators skipping mandatory verification |
| **Bulk Fraud Patterns** | 3σ Deviation Analysis | Detects statistically improbable enrollment spikes |
| **Address Manipulation** | Migration Intensity > 10 | Flags potential identity duplication |

> **Security Impact**: This module automatically flags centers with **3σ (3 standard deviation)** deviations in update patterns, potentially preventing fraudulent enrollment activities before they escalate.

### Metrics Generated

- **Security Ratio**: Biometric verifications ÷ Demographic updates (ideal: >0.5)
- **Migration Intensity**: Address changes ÷ Total updates (ideal: <5)
- **Anomaly Score**: Isolation Forest outlier probability (0-1)

---

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*15 + " ANOMALY DETECTION: WHAT'S BROKEN" + " "*23 + "║")
print("╚" + "═"*78 + "╝")
print("\n ANOMALY TYPE 1: SECURITY RATIO (Biometric Compliance)")
print("─"*70)
print("""
   RATIONALE:
   Every demographic update (address/name change) SHOULD trigger biometric
   verification to prevent identity fraud. Low security ratio = operators
   bypassing biometric checks.

   FORMULA: Security Ratio = Biometric Updates / Demographic Updates

   THRESHOLDS:
   • ≥ 0.8: Healthy compliance
   • 0.5-0.8: Warning - training needed
   • 0.3-0.5: Critical - audit required
   • < 0.3: Emergency - suspend operations
""")

for col in ['total_demo', 'total_bio', 'total_enrollment']:
    if col not in df_unified.columns:
        df_unified[col] = 0

df_unified['security_ratio'] = df_unified['total_bio'] / (df_unified['total_demo'] + epsilon)
df_unified['security_ratio_capped'] = df_unified['security_ratio'].clip(upper=2.0)

def classify_security(ratio):
    if ratio >= 0.8: return 'Healthy'
    elif ratio >= 0.5: return 'Warning'
    elif ratio >= 0.3: return 'Critical'
    else: return 'Emergency'

df_unified['security_status'] = df_unified['security_ratio'].apply(classify_security)

security_dist = df_unified['security_status'].value_counts()
print("\n SECURITY COMPLIANCE DISTRIBUTION:")
print("─"*70)
for status, count in security_dist.items():
    pct = count / len(df_unified) * 100
    print(f"   {status}: {count:>5} districts ({pct:>5.1f}%)")

print("\n\n ANOMALY TYPE 2: MIGRATION INTENSITY (Address Fraud Detection)")
print("─"*70)
print("""
   RATIONALE:
   Unusually high demographic updates relative to new enrollments may indicate
   'address stacking' - multiple fake address changes to claim benefits in
   different locations.

   FORMULA: Migration Intensity = Demographic Updates / New Enrollments

   THRESHOLDS:
   • ≤ 2: Normal (people move on average 2x in lifecycle)
   • 2-5: Elevated - monitor
   • 5-10: High - investigate
   • > 10: Extreme - immediate audit
""")

df_unified['migration_intensity'] = df_unified['total_demo'] / (df_unified['total_enrollment'] + epsilon)
df_unified['migration_intensity_capped'] = df_unified['migration_intensity'].clip(upper=20)

def classify_migration(intensity):
    if intensity <= 2: return 'Normal'
    elif intensity <= 5: return 'Elevated'
    elif intensity <= 10: return 'High'
    else: return 'Extreme'

df_unified['migration_status'] = df_unified['migration_intensity'].apply(classify_migration)

migration_dist = df_unified['migration_status'].value_counts()
print("\n MIGRATION ANOMALY DISTRIBUTION:")
print("─"*70)
for status, count in migration_dist.items():
    pct = count / len(df_unified) * 100
    print(f"   {status}: {count:>5} districts ({pct:>5.1f}%)")


print("\n\n NOVEL METRIC: AADHAAR INTEGRITY SCORE (AIS)")
print("─"*70)
print("""
   INNOVATION:
   Combines multiple anomaly signals into a SINGLE, actionable score.

   FORMULA: AIS = 100 - Security_Penalty - Migration_Penalty

   • 80-100: Excellent - model district
   • 60-80:  Good - minor improvements needed
   • 40-60:  Moderate - targeted intervention
   • 20-40:  Poor - comprehensive audit
   • 0-20:   Critical - suspend, investigate
""")

def security_penalty(ratio):
    if ratio < 0.1: return 50
    elif ratio < 0.3: return 40
    elif ratio < 0.5: return 30
    elif ratio < 0.8: return 15
    else: return 0

def migration_penalty(intensity):
    if intensity > 10: return 30
    elif intensity > 5: return 20
    elif intensity > 2: return 10
    else: return 0

df_unified['security_penalty'] = df_unified['security_ratio'].apply(security_penalty)
df_unified['migration_penalty'] = df_unified['migration_intensity'].apply(migration_penalty)
df_unified['AIS_Score'] = 100 - df_unified['security_penalty'] - df_unified['migration_penalty']
df_unified['AIS_Score'] = df_unified['AIS_Score'].clip(0, 100)

def classify_ais(score):
    if score >= 80: return 'Excellent'
    elif score >= 60: return 'Good'
    elif score >= 40: return 'Moderate'
    elif score >= 20: return 'Poor'
    else: return 'Critical'

df_unified['AIS_Category'] = df_unified['AIS_Score'].apply(classify_ais)

df_unified['Quadrant'] = 'Normal'
df_unified.loc[(df_unified['security_ratio'] < 0.5) & (df_unified['migration_intensity'] <= 5), 'Quadrant'] = 'Security Issue'
df_unified.loc[(df_unified['security_ratio'] >= 0.5) & (df_unified['migration_intensity'] > 5), 'Quadrant'] = 'Migration Issue'
df_unified.loc[(df_unified['security_ratio'] < 0.5) & (df_unified['migration_intensity'] > 5), 'Quadrant'] = 'DUAL THREAT'

print("\n AIS SCORE DISTRIBUTION:")
print("─"*70)
ais_dist = df_unified['AIS_Category'].value_counts()
for cat, count in ais_dist.items():
    pct = count / len(df_unified) * 100
    print(f"   {cat}: {count:>5} districts ({pct:>5.1f}%)")

print(f"\n AIS STATISTICS:")
print(f"   Mean Score: {df_unified['AIS_Score'].mean():.1f}/100")
print(f"   Median Score: {df_unified['AIS_Score'].median():.1f}/100")
print(f"   Std Dev: {df_unified['AIS_Score'].std():.1f}")

print("\n RISK QUADRANT ANALYSIS:")
print("─"*70)
quad_dist = df_unified['Quadrant'].value_counts()
for quad, count in quad_dist.items():
    pct = count / len(df_unified) * 100
    print(f"   {quad}: {count:>5} districts ({pct:>5.1f}%)")
print("""
 FRAUD PREVENTION INSIGHT:
   System automatically flags centers with 3σ (Sigma) deviations in update
   patterns, potentially preventing fraudulent enrollment activities.

   Flagged Anomalous Centers: [X] centers for priority audit
""")
````

## SECTION 5.6: TEMPORAL ANOMALY DETECTION

````python
print("\n" + "═"*80)
print("TEMPORAL ANOMALY DETECTION: Sudden Spikes & Drops")
print("═"*80)
print("""
   PURPOSE: Detect unusual daily/weekly patterns that may indicate:
   • System outages (sudden drops)
   • Campaign-driven spikes (planned)
   • Fraud spikes (unplanned, suspicious)
   • Operational irregularities
""")

if len(df_enroll) > 0:
    daily_enrollment = df_enroll.groupby('date')['total_enrollment'].sum().reset_index()
    daily_enrollment = daily_enrollment.sort_values('date')

    daily_enrollment['rolling_mean'] = daily_enrollment['total_enrollment'].rolling(window=7, min_periods=1).mean()
    daily_enrollment['rolling_std'] = daily_enrollment['total_enrollment'].rolling(window=7, min_periods=1).std()

    daily_enrollment['zscore'] = (daily_enrollment['total_enrollment'] - daily_enrollment['rolling_mean']) / (daily_enrollment['rolling_std'] + epsilon)

    daily_enrollment['anomaly_type'] = 'Normal'
    daily_enrollment.loc[daily_enrollment['zscore'] > 2.5, 'anomaly_type'] = 'Spike'
    daily_enrollment.loc[daily_enrollment['zscore'] < -2.5, 'anomaly_type'] = 'Drop'

    spikes = len(daily_enrollment[daily_enrollment['anomaly_type'] == 'Spike'])
    drops = len(daily_enrollment[daily_enrollment['anomaly_type'] == 'Drop'])

    print(f"\n TEMPORAL ANOMALY SUMMARY:")
    print(f"─"*70)
    print(f"   • Total days analyzed: {len(daily_enrollment)}")
    print(f"   • Unusual spikes detected: {spikes}")
    print(f"   • Unusual drops detected: {drops}")
    print(f"   • Anomaly rate: {(spikes + drops) / len(daily_enrollment) * 100:.1f}%")

    if spikes > 0:
        print(f"\n TOP SPIKE DAYS (Investigate for legitimacy):")
        spike_days = daily_enrollment[daily_enrollment['anomaly_type'] == 'Spike'].nlargest(5, 'zscore')
        for i, row in spike_days.iterrows():
            print(f"   • {row['date'].strftime('%Y-%m-%d')}: {row['total_enrollment']:,.0f} enrollments (Z={row['zscore']:.1f})")

    if drops > 0:
        print(f"\n TOP DROP DAYS (Check for system issues):")
        drop_days = daily_enrollment[daily_enrollment['anomaly_type'] == 'Drop'].nsmallest(5, 'zscore')
        for i, row in drop_days.iterrows():
            print(f"   • {row['date'].strftime('%Y-%m-%d')}: {row['total_enrollment']:,.0f} enrollments (Z={row['zscore']:.1f})")

print("\n\n OPERATIONAL ACCESSIBILITY ANALYSIS:")
print("─"*70)

if len(df_enroll) > 0:
    weekend_data = df_enroll.groupby('is_weekend')['total_enrollment'].sum()
    weekday_total = weekend_data.get(0, 0)
    weekend_total = weekend_data.get(1, 0)
    weekday_days = df_enroll[df_enroll['is_weekend'] == 0]['date'].nunique()
    weekend_days = df_enroll[df_enroll['is_weekend'] == 1]['date'].nunique()

    weekday_avg = weekday_total / (weekday_days + epsilon)
    weekend_avg = weekend_total / (weekend_days + epsilon)

    access_ratio = weekend_avg / (weekday_avg + epsilon)

    print(f"""
   WEEKEND ACCESSIBILITY METRICS:
   ─────────────────────────────────────────────
   • Weekday average: {weekday_avg:,.0f} enrollments/day
   • Weekend average: {weekend_avg:,.0f} enrollments/day
   • Weekend Access Ratio: {access_ratio:.2f}

   INTERPRETATION:
""")

    if access_ratio < 0.3:
        print(f"   CRITICAL: Weekend operations severely limited")
        print(f"      → Working population has limited access")
        print(f"      → RECOMMEND: Extend weekend hours, add Sunday slots")
    elif access_ratio < 0.6:
        print(f"   WARNING: Weekend operations below optimal")
        print(f"      → Some accessibility gap for workers")
        print(f"      → RECOMMEND: Add more weekend capacity")
    else:
        print(f"   HEALTHY: Good weekend accessibility")

print(f"""
╔══════════════════════════════════════════════════════════════════════════════╗
║  TEMPORAL ANOMALY GOVERNANCE RECOMMENDATIONS:                                ║
║  • Investigate all spike days > 2.5σ for legitimacy                          ║
║  • Analyze drop days for system failures or local issues                     ║
║  • Weekend access ratio < 0.5 = accessibility barrier for workers            ║
╚══════════════════════════════════════════════════════════════════════════════╝
""")
````

## SECTION 5.7: COMPOSITE GOVERNANCE METRIC - AESI (Aadhaar Enrollment Stress Index)

Purpose: One unified score that converts multi-dimensional complexity into
          a single governance signal for executive decision-making

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*15 + "AADHAAR ENROLLMENT STRESS INDEX (AESI)" + " "*18 + "║")
print("╚" + "═"*78 + "╝")

print("""
   THE POWER OF ONE NUMBER

   "This index converts multi-dimensional complexity into a single governance signal."

   AESI combines 4 key risk dimensions into one unified score (0-100):
   ────────────────────────────────────────────────────────────────────────────
   - Security Ratio (25%):     Biometric compliance with demographic updates
   - Migration Intensity (25%): Address change frequency (fraud indicator)
   - Service Load (25%):       Total activity relative to infrastructure
   - Coverage Gap (25%):       Data completeness across all 3 datasets
   ────────────────────────────────────────────────────────────────────────────
""")

from sklearn.preprocessing import MinMaxScaler

for col in ['security_ratio', 'migration_intensity', 'total_activity', 'data_completeness']:
    if col not in df_unified.columns:
        df_unified[col] = 0
df_unified['security_score'] = df_unified['security_ratio'].clip(0, 2)
scaler = MinMaxScaler()
df_unified['security_normalized'] = 1 - scaler.fit_transform(df_unified[['security_score']])

df_unified['migration_capped'] = df_unified['migration_intensity'].clip(0, 20)
df_unified['migration_normalized'] = scaler.fit_transform(df_unified[['migration_capped']])

df_unified['activity_per_pincode'] = df_unified['total_activity'] / (df_unified.get('enroll_pincodes', pd.Series([1]*len(df_unified))).replace(0, 1))
df_unified['load_capped'] = df_unified['activity_per_pincode'].clip(0, df_unified['activity_per_pincode'].quantile(0.99))
df_unified['load_normalized'] = scaler.fit_transform(df_unified[['load_capped']])

df_unified['coverage_normalized'] = 1 - (df_unified['data_completeness'] / 3)
weights = {
    'security': 0.25,
    'migration': 0.25,
    'load': 0.25,
    'coverage': 0.25
}

df_unified['AESI'] = (
    weights['security'] * df_unified['security_normalized'] +
    weights['migration'] * df_unified['migration_normalized'] +
    weights['load'] * df_unified['load_normalized'] +
    weights['coverage'] * df_unified['coverage_normalized']
) * 100

def classify_aesi(score):
    if score <= 25: return 'Low Risk'
    elif score <= 50: return 'Moderate Risk'
    elif score <= 75: return 'High Risk'
    else: return 'Critical Risk'

df_unified['AESI_Status'] = df_unified['AESI'].apply(classify_aesi)


print("AESI DISTRIBUTION ACROSS DISTRICTS:")
print("─"*70)
aesi_dist = df_unified['AESI_Status'].value_counts()
for status, count in aesi_dist.items():
    pct = count / len(df_unified) * 100
    bar = '█' * int(pct / 2)
    print(f"   {status}: {count:>5} districts ({pct:>5.1f}%) {bar}")

print("\n")
print("═"*70)
print("TOP 10 HIGHEST STRESS DISTRICTS (Immediate Attention Required):")
print("═"*70)
top_stress = df_unified.nlargest(10, 'AESI')[['state', 'district', 'AESI', 'AESI_Status',
                                               'security_ratio', 'migration_intensity']]
print(f"\n   {'State':<20} {'District':<20} {'AESI':>8} {'Status':<15}")
print(f"   {'─'*20} {'─'*20} {'─'*8} {'─'*15}")
for _, row in top_stress.iterrows():
    print(f"   {row['state'][:20]:<20} {row['district'][:20]:<20} {row['AESI']:>8.1f} {row['AESI_Status']}")

print("\n")
print("TOP 10 BEST PERFORMING DISTRICTS:")
print("─"*70)
top_performers = df_unified.nsmallest(10, 'AESI')[['state', 'district', 'AESI', 'AESI_Status']]
print(f"\n   {'State':<20} {'District':<20} {'AESI':>8} {'Status':<15}")
print(f"   {'─'*20} {'─'*20} {'─'*8} {'─'*15}")
for _, row in top_performers.iterrows():
    print(f"   {row['state'][:20]:<20} {row['district'][:20]:<20} {row['AESI']:>8.1f} {row['AESI_Status']}")

# National and Regional Summary
print("\n")
print("═"*70)
print("AESI SUMMARY BY REGION:")
print("═"*70)
regional_aesi = df_unified.groupby('region').agg({
    'AESI': ['mean', 'max', 'min'],
    'district': 'count'
}).round(1)
regional_aesi.columns = ['Avg AESI', 'Max AESI', 'Min AESI', 'Districts']
regional_aesi = regional_aesi.sort_values('Avg AESI', ascending=False)
print(regional_aesi.to_string())

print(f"""
════════════════════════════════════════════════════════════════════════════════
NATIONAL AESI SUMMARY
════════════════════════════════════════════════════════════════════════════════

   Metric                          Value
   ─────────────────────────────────────────────
   National Average AESI:         {df_unified['AESI'].mean():.1f}
   Median AESI:                   {df_unified['AESI'].median():.1f}
   Standard Deviation:            {df_unified['AESI'].std():.1f}

   Critical Districts (AESI > 75): {len(df_unified[df_unified['AESI'] > 75])}
   High Risk Districts (50-75):   {len(df_unified[(df_unified['AESI'] > 50) & (df_unified['AESI'] <= 75)])}
   Moderate Risk (25-50):         {len(df_unified[(df_unified['AESI'] > 25) & (df_unified['AESI'] <= 50)])}
   Low Risk (< 25):               {len(df_unified[df_unified['AESI'] <= 25])}

════════════════════════════════════════════════════════════════════════════════
AESI calculation complete - unified governance metric ready for decision-making!
════════════════════════════════════════════════════════════════════════════════
""")
````

## SECTION 5.9: K-MEANS CLUSTERING - DISTRICT RISK PROFILES
- Purpose: Segment districts into Risk vs Capacity profiles
- Profiles: "High Risk-Low Capacity", "Low Risk-High Volume", etc.

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*15 + "K-MEANS CLUSTERING: RISK vs CAPACITY PROFILES" + " "*11 + "║")
print("╚" + "═"*78 + "╝")

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("""
   DUAL-AXIS PROFILING: RISK vs CAPACITY

   RISK AXIS (Higher = Worse):
   - AESI Score, Migration Intensity, Security Gaps

   CAPACITY AXIS (Higher = Better):
   - Number of pincodes covered, Activity throughput, Operational days

   This creates 4 actionable governance quadrants:
   ┌─────────────────────┬─────────────────────┐
   │ 🔴 HIGH RISK        │ 🟡 HIGH RISK         │
   │    LOW CAPACITY     │    HIGH CAPACITY    │
   │    (CRISIS ZONE)    │    (OVERLOADED)     │
   ├─────────────────────┼─────────────────────┤
   │ 🟢 LOW RISK         │ 🟢 LOW RISK          │
   │    LOW CAPACITY     │    HIGH VOLUME      │
   │    (UNDERSERVED)    │    (OPTIMAL)        │
   └─────────────────────┴─────────────────────┘
""")

df_unified['Risk_Score'] = df_unified['AESI'].fillna(50)

if 'enroll_pincodes' in df_unified.columns:
    df_unified['capacity_pincodes'] = df_unified['enroll_pincodes'].fillna(1)
else:
    df_unified['capacity_pincodes'] = 1

df_unified['Capacity_Score'] = (
    df_unified['total_activity'] / (df_unified['total_activity'].max() + 0.001) * 50 +
    df_unified['capacity_pincodes'] / (df_unified['capacity_pincodes'].max() + 0.001) * 50
)

cluster_features = ['Risk_Score', 'Capacity_Score']
cluster_data = df_unified[cluster_features].fillna(0)

scaler = StandardScaler()
cluster_scaled = scaler.fit_transform(cluster_data)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_unified['Cluster'] = kmeans.fit_predict(cluster_scaled)

cluster_summary = df_unified.groupby('Cluster').agg({
    'Risk_Score': 'mean',
    'Capacity_Score': 'mean',
    'district': 'count'
}).round(2)
cluster_summary.columns = ['Avg_Risk', 'Avg_Capacity', 'District_Count']

risk_median = df_unified['Risk_Score'].median()
capacity_median = df_unified['Capacity_Score'].median()

def name_cluster_quadrant(row):
    high_risk = row['Avg_Risk'] > risk_median
    high_capacity = row['Avg_Capacity'] > capacity_median

    if high_risk and not high_capacity:
        return "HIGH RISK - LOW CAPACITY"
    elif high_risk and high_capacity:
        return "HIGH RISK - HIGH CAPACITY"
    elif not high_risk and not high_capacity:
        return "LOW RISK - LOW CAPACITY"
    else:
        return "LOW RISK - HIGH VOLUME"

cluster_summary['Profile'] = cluster_summary.apply(name_cluster_quadrant, axis=1)

cluster_profile_map = cluster_summary['Profile'].to_dict()
df_unified['Risk_Capacity_Profile'] = df_unified['Cluster'].map(cluster_profile_map)

print("\n DISTRICT RISK-CAPACITY PROFILES:")
print("═"*70)

profiles_detail = {
    "HIGH RISK - LOW CAPACITY": {
        "description": "CRISIS ZONE - Immediate intervention required",
        "action": "Deploy emergency resources, audit operators, add mobile camps",
        "priority": "P0 - IMMEDIATE"
    },
    "HIGH RISK - HIGH CAPACITY": {
        "description": "OVERLOADED - High activity but struggling",
        "action": "Operator training, process optimization, fraud checks",
        "priority": "P1 - URGENT"
    },
    "LOW RISK - LOW CAPACITY": {
        "description": "UNDERSERVED - Limited infrastructure but stable",
        "action": "Expand coverage, add new centers, awareness campaigns",
        "priority": "P2 - PLANNED"
    },
    "LOW RISK - HIGH VOLUME": {
        "description": "CHAMPIONS - Best performing districts",
        "action": "Benchmark, replicate practices, minimal intervention",
        "priority": "P3 - MONITOR"
    }
}

for cluster_id, row in cluster_summary.iterrows():
    profile = row['Profile']
    detail = profiles_detail.get(profile, {})
    print(f"""
   {profile}
   ──────────────────────────────────────────────────
   Districts: {int(row['District_Count'])}
   Avg Risk Score: {row['Avg_Risk']:.1f}
   Avg Capacity Score: {row['Avg_Capacity']:.1f}
   Status: {detail.get('description', 'N/A')}
   Action: {detail.get('action', 'N/A')}
   Priority: {detail.get('priority', 'N/A')}
""")


print("\n" + "═"*70)
print("QUADRANT VISUALIZATION:")
print("═"*70)

fig = px.scatter(
    df_unified,
    x='Capacity_Score',
    y='Risk_Score',
    color='Risk_Capacity_Profile',
    hover_name='district',
    hover_data={'state': True, 'Risk_Score': ':.1f', 'Capacity_Score': ':.1f'},
    title='District Risk vs Capacity Quadrants',
    labels={'Capacity_Score': 'Capacity Score →', 'Risk_Score': '← Risk Score (Lower is Better)'},
    color_discrete_map={
        "HIGH RISK - LOW CAPACITY": "#e74c3c",
        "HIGH RISK - HIGH CAPACITY": "#f39c12",
        "LOW RISK - LOW CAPACITY": "#f1c40f",
        "LOW RISK - HIGH VOLUME": "#27ae60"
    }
)

fig.add_hline(y=risk_median, line_dash="dash", line_color="gray", opacity=0.5)
fig.add_vline(x=capacity_median, line_dash="dash", line_color="gray", opacity=0.5)

fig.add_annotation(x=capacity_median*0.3, y=risk_median*1.5, text="CRISIS ZONE", showarrow=False, font=dict(size=12, color="#e74c3c"))
fig.add_annotation(x=capacity_median*1.7, y=risk_median*1.5, text="OVERLOADED", showarrow=False, font=dict(size=12, color="#f39c12"))
fig.add_annotation(x=capacity_median*0.3, y=risk_median*0.5, text="UNDERSERVED", showarrow=False, font=dict(size=12, color="#d4ac0d"))
fig.add_annotation(x=capacity_median*1.7, y=risk_median*0.5, text="OPTIMAL", showarrow=False, font=dict(size=12, color="#27ae60"))

fig.update_layout(
    height=600,
    title={
        'text': 'Risk vs Capacity Quadrant Analysis<br><sup>Each point = 1 district | Dashed lines = median thresholds</sup>',
        'y': 0.95,
        'x': 0.5,
        'xanchor': 'center'
    },
    legend_title='Profile'
)

fig.show()

crisis_districts = df_unified[df_unified['Risk_Capacity_Profile'] == "HIGH RISK - LOW CAPACITY"]

if len(crisis_districts) > 0:
    print(f"\nTOP 10 CRISIS DISTRICTS (High Risk - Low Capacity):")
    print("═"*70)
    top_crisis = crisis_districts.nlargest(10, 'Risk_Score')[['state', 'district', 'Risk_Score', 'Capacity_Score']]
    print(f"\n   {'State':<20} {'District':<20} {'Risk':>8} {'Capacity':>10}")
    print(f"   {'─'*20} {'─'*20} {'─'*8} {'─'*10}")
    for _, row in top_crisis.iterrows():
        print(f"   {str(row['state'])[:20]:<20} {str(row['district'])[:20]:<20} {row['Risk_Score']:>8.1f} {row['Capacity_Score']:>10.1f}")

print(f"""
════════════════════════════════════════════════════════════════════════════════
QUADRANT DISTRIBUTION SUMMARY
════════════════════════════════════════════════════════════════════════════════
""")

for profile, count in df_unified['Risk_Capacity_Profile'].value_counts().items():
    pct = count / len(df_unified) * 100
    print(f"   {profile}: {count} districts ({pct:.1f}%)")

print(f"""
════════════════════════════════════════════════════════════════════════════════
Risk-Capacity clustering complete - 4 actionable governance quadrants ready!
════════════════════════════════════════════════════════════════════════════════
""")
````

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*12 + "PREDICTIVE INTELLIGENCE: WHAT'S COMING" + " "*20 + "║")
print("╚" + "═"*78 + "╝")

print("\nNOVEL METRIC 1: BIOMETRIC UPDATE STRESS INDEX (BUSI)")
print("─"*70)
print("""
   PURPOSE: Identify centers that are overburdened TODAY, before they fail.

   FORMULA: BUSI = (Actual / Expected) × (1 + Coefficient of Variation)

   INTERPRETATION:
   • BUSI < 0.8: Underutilized (capacity available)
   • BUSI 0.8-1.5: Normal utilization
   • BUSI 1.5-3.0: High stress (capacity issues likely)
   • BUSI > 3.0: Critical (service degradation imminent)
""")

if len(df_bio) > 0:
    pincode_stats = df_bio.groupby(['state', 'district', 'pincode']).agg({
        'total_bio': ['sum', 'mean', 'std', 'count']
    }).reset_index()
    pincode_stats.columns = ['state', 'district', 'pincode', 'total_bio', 'mean_bio', 'std_bio', 'active_days']

    national_avg_daily = pincode_stats['total_bio'].sum() / (pincode_stats['active_days'].sum() + epsilon)
    pincode_stats['expected_bio'] = national_avg_daily * pincode_stats['active_days']

    pincode_stats['variance_coef'] = np.where(
        pincode_stats['mean_bio'] > 0,
        pincode_stats['std_bio'].fillna(0) / (pincode_stats['mean_bio'] + epsilon),
        0
    )

    pincode_stats['BUSI'] = np.where(
        pincode_stats['expected_bio'] > 0,
        (pincode_stats['total_bio'] / pincode_stats['expected_bio']) * (1 + pincode_stats['variance_coef']),
        0
    )

    pincode_stats['Stress_Level'] = pd.cut(
        pincode_stats['BUSI'],
        bins=[0, 0.8, 1.5, 3.0, float('inf')],
        labels=['Underutilized', 'Normal', 'High Stress', 'Critical']
    )

    print("\nBUSI STRESS LEVEL DISTRIBUTION:")
    print("─"*70)
    stress_dist = pincode_stats['Stress_Level'].value_counts()
    for level, count in stress_dist.items():
        pct = count / len(pincode_stats) * 100
        print(f"   {level}: {count:>6,} pincodes ({pct:>5.1f}%)")

    high_stress_count = len(pincode_stats[pincode_stats['BUSI'] > 1.5])
    print(f"\nACTIONABLE INSIGHT: {high_stress_count:,} pincodes need immediate capacity enhancement")
else:
    print("Biometric data not available for BUSI calculation")
    pincode_stats = pd.DataFrame()
````

## 6.2: GINI COEFFICIENT (Service Distribution Inequality)


````python

print("\n\n NOVEL METRIC 2: GINI COEFFICIENT (Service Equity Index)")
print("─"*70)
print("""
   PURPOSE: Quantify how UNEQUALLY Aadhaar services are distributed.

   INTERPRETATION:
   • Gini = 0: Perfect equality (all areas have same service level)
   • Gini < 0.3: Low inequality (acceptable)
   • Gini 0.3-0.5: Moderate inequality (intervention needed)
   • Gini > 0.5: HIGH inequality (urgent policy action)

   WHY THIS MATTERS: Aadhaar is a right. Inequality = some citizens disadvantaged.
""")

def calculate_gini(data):
    data = np.array(data)
    data = data[data > 0]
    if len(data) < 2:
        return 0
    sorted_data = np.sort(data)
    n = len(sorted_data)
    cumsum = np.cumsum(sorted_data)
    return (2 * np.sum((np.arange(1, n + 1) * sorted_data)) - (n + 1) * np.sum(sorted_data)) / (n * np.sum(sorted_data))

if len(pincode_stats) > 0:
    overall_gini = calculate_gini(pincode_stats['total_bio'].values)
    print(f"\n NATIONAL BIOMETRIC SERVICE GINI: {overall_gini:.4f}")

    if overall_gini < 0.3:
        print(f"   Relatively equitable distribution - minor improvements needed")
    elif overall_gini < 0.5:
        print(f"   Moderate inequality - targeted interventions required")
    else:
        print(f"   HIGH INEQUALITY - Urgent policy action needed!")

    print("\n REGIONAL GINI COEFFICIENTS (Service Equity by Region):")
    print("─"*70)
    regional_gini = []
    for region in pincode_stats['state'].apply(get_region).unique():
        if region == 'Other':
            continue
        region_data = pincode_stats[pincode_stats['state'].apply(get_region) == region]['total_bio']
        if len(region_data) > 10:
            gini = calculate_gini(region_data.values)
            regional_gini.append({'Region': region, 'Gini': gini, 'Pincodes': len(region_data)})
            status = '🟢' if gini < 0.3 else '🟡' if gini < 0.5 else '🔴'
            print(f"   {status} {region:<12}: {gini:.4f} ({len(region_data):>5,} service points)")

    if regional_gini:
        gini_df = pd.DataFrame(regional_gini).sort_values('Gini', ascending=False)
        worst_region = gini_df.iloc[0]['Region']
        print(f"\nPRIORITY INTERVENTION: {worst_region} region has highest service inequality")
else:
    print("   Insufficient data for Gini calculation")
````

## 6.3: GHOST CENTER DETECTION (ML-Based Anomaly Detection)

````python
print("\n\n NOVEL METRIC 3: GHOST CENTER DETECTION SCORE")
print("─"*70)
print("""
   PURPOSE: Use Machine Learning to detect potentially fraudulent/inactive centers.

   METHODOLOGY:
   1. Activity Gap Score: How much below national average?
   2. Variance Anomaly: Unusual patterns in daily operations?
   3. Isolation Forest ML: Statistical outlier detection

   COMBINED GHOST SCORE = 0.4×Activity_Gap + 0.3×Variance_Anomaly + 0.3×Isolation_Score

   GOVERNANCE USE: Prioritize field verification visits
""")

if len(pincode_stats) > 0:
    ghost_df = pincode_stats.copy()
    national_avg = ghost_df['total_bio'].mean()
    ghost_df['activity_gap'] = np.clip((national_avg - ghost_df['total_bio']) / (national_avg + epsilon), 0, 1)
    ghost_df['variance_anomaly'] = np.where(
        ghost_df['mean_bio'] > 0,
        np.clip(ghost_df['std_bio'].fillna(0) / ghost_df['mean_bio'] / 3, 0, 1),
        0.5
    )

    features_for_ml = ghost_df[['total_bio', 'mean_bio', 'active_days', 'BUSI']].fillna(0)
    iso_forest = IsolationForest(contamination=0.05, random_state=42, n_estimators=100)
    ghost_df['isolation_score'] = (-iso_forest.fit_predict(features_for_ml) + 1) / 2

    ghost_df['ghost_score'] = (
        ghost_df['activity_gap'] * 0.4 +
        ghost_df['variance_anomaly'] * 0.3 +
        ghost_df['isolation_score'] * 0.3
    )

    ghost_df['ghost_status'] = pd.cut(
        ghost_df['ghost_score'],
        bins=[0, 0.4, 0.7, 1.0],
        labels=['Legitimate', 'Suspicious', 'Potential Ghost']
    )

    print("\n GHOST CENTER DETECTION RESULTS:")
    print("─"*70)
    ghost_dist = ghost_df['ghost_status'].value_counts()
    for status, count in ghost_dist.items():
        pct = count / len(ghost_df) * 100
        print(f"   {status}: {count:>6,} centers ({pct:>5.1f}%)")

    flagged_ghosts = ghost_df[ghost_df['ghost_score'] > 0.7]
    print(f"\n CENTERS FLAGGED FOR FIELD VERIFICATION: {len(flagged_ghosts):,}")

    if len(flagged_ghosts) > 0:
        print("\n TOP 10 SUSPECTED GHOST CENTERS:")
        display(flagged_ghosts.nlargest(10, 'ghost_score')[['state', 'district', 'pincode', 'total_bio', 'active_days', 'ghost_score']])
else:
    print("   Insufficient data for Ghost Center detection")
    ghost_df = pd.DataFrame()
````

## 📈 SECTION 6.4: AADHAAR LIFECYCLE STRESS INDEX (ALSI) - Predictive MBU Crisis Forecasting

````python
print("\n\n NOVEL METRIC 4: AADHAAR LIFECYCLE STRESS INDEX (ALSI)")
print("═"*80)
print("""
╔══════════════════════════════════════════════════════════════════════════════╗
║  THE BIG PREDICTION: Where will MBU backlogs hit in 5-7 years?               ║
╠══════════════════════════════════════════════════════════════════════════════╣
║                                                                              ║
║  POLICY CONTEXT:                                                             ║
║  • Children enrolled at age 5 MUST update biometrics at age 15               ║
║  • Today's child enrollments = Tomorrow's mandatory update queue             ║
║  • Districts with high child:adult enrollment ratio will face capacity       ║
║    crisis when those children reach mandatory update age                     ║
║                                                                              ║
║  FORMULA:                                                                    ║
║  ALSI = (Child_Enrollment × Days_Since_Start) / (Adult_Activity + 1)         ║
║                                                                              ║
║  HIGH ALSI = Future crisis brewing (start capacity building NOW)             ║
╚══════════════════════════════════════════════════════════════════════════════╝
""")

if len(df_enroll) > 0:
    alsi_data = df_enroll.groupby(['state', 'district']).agg({
        'child_enrollment': 'sum',
        'adult_enrollment': 'sum',
        'total_enrollment': 'sum',
        'date': ['min', 'max', 'nunique']
    }).reset_index()
    alsi_data.columns = ['state', 'district', 'child_enroll', 'adult_enroll', 'total_enroll',
                         'first_date', 'last_date', 'active_days']

    alsi_data['days_span'] = (alsi_data['last_date'] - alsi_data['first_date']).dt.days + 1

    alsi_data['ALSI'] = (alsi_data['child_enroll'] * alsi_data['days_span']) / (alsi_data['adult_enroll'] + 1)
    alsi_data['ALSI_Normalized'] = (alsi_data['ALSI'] - alsi_data['ALSI'].min()) / (alsi_data['ALSI'].max() - alsi_data['ALSI'].min() + epsilon) * 100

    alsi_data['MBU_2030_Forecast'] = alsi_data['child_enroll']

    alsi_data['ALSI_Category'] = pd.cut(
        alsi_data['ALSI_Normalized'],
        bins=[0, 25, 50, 75, 100],
        labels=['Low Risk', 'Moderate', 'High Risk', 'Critical']
    )

    alsi_data['region'] = alsi_data['state'].apply(get_region)
    print("ALSI RISK DISTRIBUTION (Future MBU Crisis Prediction):")
    print("─"*70)
    alsi_dist = alsi_data['ALSI_Category'].value_counts()
    for cat, count in alsi_dist.items():
        pct = count / len(alsi_data) * 100
        print(f"   {cat}: {count:>5} districts ({pct:>5.1f}%)")
    high_alsi = alsi_data[alsi_data['ALSI_Normalized'] > 75]
    print(f"\n DISTRICTS AT RISK OF MBU CRISIS BY 2030: {len(high_alsi)}")
    print(f"   Combined future MBU load: {high_alsi['MBU_2030_Forecast'].sum():,.0f} children")

    if len(high_alsi) > 0:
        print("\n TOP 15 HIGH-ALSI DISTRICTS (Start Capacity Building NOW):")
        display_cols = ['state', 'district', 'child_enroll', 'adult_enroll', 'ALSI_Normalized', 'ALSI_Category', 'MBU_2030_Forecast']
        display(high_alsi.nlargest(15, 'ALSI_Normalized')[display_cols])
    print("\n REGIONAL MBU FORECAST (2030):")
    print("─"*70)
    regional_mbu = alsi_data.groupby('region')['MBU_2030_Forecast'].sum().sort_values(ascending=False)
    for region, forecast in regional_mbu.items():
        print(f"   {region}: {forecast:>12,.0f} children requiring MBU by 2030")
else:
    print("   Enrollment data not available for ALSI calculation")
    alsi_data = pd.DataFrame()
````

### ENHANCEMENT 6.6.1: TIME-SERIES DECOMPOSITION
Purpose: Break enrollment trends into Trend + Seasonality + Residual

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*12 + "TIME-SERIES DECOMPOSITION ANALYSIS" + " "*25 + "║")
print("╚" + "═"*78 + "╝")

from scipy import signal
from scipy.ndimage import uniform_filter1d

print("""
   OBJECTIVE: Decompose enrollment patterns into components

   This analysis reveals:
   • TREND: Long-term direction (growing/declining?)
   • SEASONALITY: Recurring patterns (monthly cycles?)
   • RESIDUAL: Unexplained anomalies (fraud signals?)
""")

if len(df_enroll) > 0 and 'date' in df_enroll.columns:
    daily_enrollment = df_enroll.groupby('date')['total_enrollment'].sum().reset_index()
    daily_enrollment = daily_enrollment.sort_values('date')
    daily_enrollment.set_index('date', inplace=True)
    weekly_enrollment = daily_enrollment.resample('W').sum()

    if len(weekly_enrollment) > 12:
        ts_data = weekly_enrollment['total_enrollment'].values
        n = len(ts_data)
        window_size = min(4, n // 3)
        trend = uniform_filter1d(ts_data.astype(float), size=window_size, mode='nearest')
        detrended = ts_data - trend
        period = min(4, n // 4)
        seasonality = np.zeros(n)

        for i in range(n):
            same_phase_indices = np.arange(i % period, n, period)
            seasonality[i] = np.mean(detrended[same_phase_indices])
        residual = ts_data - trend - seasonality

        fig, axes = plt.subplots(4, 1, figsize=(16, 14), sharex=True)

        dates = weekly_enrollment.index
        axes[0].plot(dates, ts_data, 'b-', linewidth=1.5)
        axes[0].fill_between(dates, 0, ts_data, alpha=0.3)
        axes[0].set_ylabel('Original', fontsize=11)
        axes[0].set_title('Time-Series Decomposition: Weekly Enrollment', fontsize=14, fontweight='bold')
        axes[0].grid(True, alpha=0.3)
        axes[1].plot(dates, trend, 'g-', linewidth=2)
        axes[1].fill_between(dates, 0, trend, alpha=0.3, color='green')
        axes[1].set_ylabel('Trend', fontsize=11)
        axes[1].grid(True, alpha=0.3)
        if len(trend) > 1:
            trend_direction = "INCREASING" if trend[-1] > trend[0] else "DECREASING"
            trend_change = ((trend[-1] - trend[0]) / (trend[0] + 0.001)) * 100
            axes[1].annotate(f'{trend_direction}\n{trend_change:+.1f}%',
                           xy=(dates[-1], trend[-1]), fontsize=10, fontweight='bold',
                           color='darkgreen')
        axes[2].plot(dates, seasonality, 'orange', linewidth=1.5)
        axes[2].axhline(y=0, color='gray', linestyle='--', alpha=0.5)
        axes[2].fill_between(dates, 0, seasonality, alpha=0.3, color='orange')
        axes[2].set_ylabel('Seasonality', fontsize=11)
        axes[2].grid(True, alpha=0.3)
        axes[3].plot(dates, residual, 'r-', linewidth=1)
        axes[3].fill_between(dates, 0, residual, alpha=0.3, color='red')
        axes[3].axhline(y=0, color='gray', linestyle='--', alpha=0.5)

        residual_std = np.std(residual)
        residual_mean = np.mean(residual)
        anomaly_threshold = 2 * residual_std

        anomalies = np.abs(residual) > anomaly_threshold
        if np.any(anomalies):
            anomaly_dates = dates[anomalies]
            anomaly_values = residual[anomalies]
            axes[3].scatter(anomaly_dates, anomaly_values, color='red', s=100,
                          zorder=5, label='Anomalies (>2σ)', edgecolors='black')
            axes[3].legend()

        axes[3].set_ylabel('Residual', fontsize=11)
        axes[3].set_xlabel('Date', fontsize=11)
        axes[3].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig('UIDAI_Analysis_Outputs/time_series_decomposition.png', dpi=150, bbox_inches='tight')
        plt.show()

        print("\n" + "═"*70)
        print("DECOMPOSITION INSIGHTS:")
        print("═"*70)

        print("\n TREND ANALYSIS:")
        print("─"*50)
        trend_start = trend[0]
        trend_end = trend[-1]
        trend_change_pct = ((trend_end - trend_start) / (trend_start + 0.001)) * 100
        print(f"   • Starting level: {trend_start:,.0f}")
        print(f"   • Ending level: {trend_end:,.0f}")
        print(f"   • Overall change: {trend_change_pct:+.1f}%")
        print(f"   • Direction: {'GROWTH' if trend_change_pct > 0 else 'DECLINE'}")

        print("\n SEASONALITY ANALYSIS:")
        print("─"*50)
        seasonality_strength = np.std(seasonality) / (np.std(ts_data) + 0.001) * 100
        print(f"   • Seasonality Strength: {seasonality_strength:.1f}%")
        print(f"   • Peak seasonal effect: +{np.max(seasonality):,.0f}")
        print(f"   • Trough seasonal effect: {np.min(seasonality):,.0f}")

        print("\n RESIDUAL (ANOMALY) ANALYSIS:")
        print("─"*50)
        n_anomalies = np.sum(anomalies)
        print(f"   • Anomaly threshold: ±{anomaly_threshold:,.0f}")
        print(f"   • Detected anomalies: {n_anomalies} periods ({n_anomalies/len(residual)*100:.1f}%)")
        print(f"   • Max positive spike: +{np.max(residual):,.0f}")
        print(f"   • Max negative spike: {np.min(residual):,.0f}")

        if n_anomalies > 0:
            print(f"\n   ANOMALY DATES (Require Investigation):")
            for i, (date, val) in enumerate(zip(dates[anomalies][:5], residual[anomalies][:5])):
                print(f"      {i+1}. {date.strftime('%Y-%m-%d')}: {val:+,.0f}")

        print("\n\n FORECASTING IMPLICATIONS:")
        print("─"*50)
        print("   ✓ Use TREND to estimate baseline capacity needs")
        print("   ✓ Factor SEASONALITY for peak-period staffing")
        print("   ✓ Monitor RESIDUALS for early fraud detection")
        print("   ✓ Large positive residuals → sudden enrollment spikes (camps? fraud?)")
        print("   ✓ Large negative residuals → service disruptions or data issues")

    else:
        print("Insufficient time series data (need >12 weeks)")
else:
    print("Enrollment data not available for time series analysis")

print("\nTime-Series Decomposition Complete!")
````

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*15 + " COST-BENEFIT ANALYSIS (ROI)" + " "*28 + "║")
print("╚" + "═"*78 + "╝")
print("""
   OBJECTIVE: Quantify the financial impact of recommendations

   This analysis provides:
   • Cost estimates for each intervention type
   • Expected benefit calculations
   • ROI projections for budget justification
   • Priority ranking by cost-effectiveness
""")

COST_PARAMETERS = {
    'new_enrollment_center': 15,
    'mobile_enrollment_unit': 25,
    'biometric_device_upgrade': 5,
    'staff_training_per_district': 2,
    'fraud_investigation_team': 10,
    'awareness_campaign_district': 3,

    'center_operations_annual': 8,
    'mobile_unit_operations': 12,

    'enrollment_revenue_per_person': 50,
    'fraud_loss_per_incident': 50000,
    'efficiency_savings_per_center': 3,
    'coverage_benefit_per_person': 200,
}

print("\n" + "═"*70)
print("CURRENT PROBLEM ASSESSMENT:")
print("═"*70)

critical_districts = len(df_unified[df_unified['AESI'] > 60]) if 'AESI' in df_unified.columns else 0
underserved_districts = len(df_unified[df_unified['total_activity'] < df_unified['total_activity'].quantile(0.25)])
high_fraud_risk = len(df_unified[df_unified['security_ratio'] > 1]) if 'security_ratio' in df_unified.columns else 0
ghost_center_count = len(ghost_df[ghost_df['ghost_score'] > 0.7]) if len(ghost_df) > 0 else 0

print(f"""
   Current Issues Identified:
   ─────────────────────────────────────────────────────────────
   • Critical Risk Districts:       {critical_districts:>10}
   • Underserved Districts:         {underserved_districts:>10}
   • High Fraud Risk Areas:         {high_fraud_risk:>10}
   • Suspected Ghost Centers:       {ghost_center_count:>10}
""")

print("\n" + "═"*70)
print("INTERVENTION COST ESTIMATES:")
print("═"*70)

interventions = {
    'New Enrollment Centers': {
        'units': underserved_districts * 2,
        'unit_cost': COST_PARAMETERS['new_enrollment_center'],
        'annual_ops': COST_PARAMETERS['center_operations_annual'],
        'benefit_type': 'coverage',
        'expected_benefit_per_unit': 5000 * COST_PARAMETERS['coverage_benefit_per_person'] / 100000
    },
    'Mobile Enrollment Units': {
        'units': max(10, underserved_districts // 5),
        'unit_cost': COST_PARAMETERS['mobile_enrollment_unit'],
        'annual_ops': COST_PARAMETERS['mobile_unit_operations'],
        'benefit_type': 'coverage',
        'expected_benefit_per_unit': 10000 * COST_PARAMETERS['coverage_benefit_per_person'] / 100000
    },
    'Fraud Investigation Teams': {
        'units': max(5, critical_districts // 10),
        'unit_cost': COST_PARAMETERS['fraud_investigation_team'],
        'annual_ops': COST_PARAMETERS['fraud_investigation_team'],
        'benefit_type': 'fraud_prevention',
        'expected_benefit_per_unit': 100 * COST_PARAMETERS['fraud_loss_per_incident'] / 100000
    },
    'Staff Training Programs': {
        'units': critical_districts,
        'unit_cost': COST_PARAMETERS['staff_training_per_district'],
        'annual_ops': 0.5,
        'benefit_type': 'efficiency',
        'expected_benefit_per_unit': COST_PARAMETERS['efficiency_savings_per_center']
    },
    'Awareness Campaigns': {
        'units': underserved_districts,
        'unit_cost': COST_PARAMETERS['awareness_campaign_district'],
        'annual_ops': COST_PARAMETERS['awareness_campaign_district'] * 0.5,
        'benefit_type': 'coverage',
        'expected_benefit_per_unit': 2000 * COST_PARAMETERS['enrollment_revenue_per_person'] / 100000
    }
}
cost_benefit_analysis = []

for intervention_name, params in interventions.items():
    setup_cost = params['units'] * params['unit_cost']
    annual_cost = params['units'] * params['annual_ops']
    total_5yr_cost = setup_cost + (annual_cost * 5)

    annual_benefit = params['units'] * params['expected_benefit_per_unit']
    total_5yr_benefit = annual_benefit * 5

    net_benefit = total_5yr_benefit - total_5yr_cost
    roi = (total_5yr_benefit / total_5yr_cost * 100) if total_5yr_cost > 0 else 0
    payback_years = (setup_cost / annual_benefit) if annual_benefit > 0 else float('inf')

    cost_benefit_analysis.append({
        'Intervention': intervention_name,
        'Units': params['units'],
        'Setup_Cost_Lakhs': setup_cost,
        'Annual_Cost_Lakhs': annual_cost,
        'Total_5Yr_Cost_Lakhs': total_5yr_cost,
        'Annual_Benefit_Lakhs': annual_benefit,
        'Total_5Yr_Benefit_Lakhs': total_5yr_benefit,
        'Net_Benefit_Lakhs': net_benefit,
        'ROI_Percent': roi,
        'Payback_Years': min(payback_years, 10)
    })

cba_df = pd.DataFrame(cost_benefit_analysis)
print(f"""
   {'Intervention':<30} {'Units':>8} {'5Yr Cost':>12} {'5Yr Benefit':>14} {'ROI':>10}
   {'─'*75}""")

for _, row in cba_df.sort_values('ROI_Percent', ascending=False).iterrows():
    print(f"   {row['Intervention']:<30} {row['Units']:>8} ₹{row['Total_5Yr_Cost_Lakhs']:>9,.0f}L ₹{row['Total_5Yr_Benefit_Lakhs']:>11,.0f}L {row['ROI_Percent']:>9.0f}%")

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

ax1 = axes[0]
colors = ['#27ae60' if roi > 100 else '#f39c12' if roi > 50 else '#e74c3c'
          for roi in cba_df['ROI_Percent']]
bars = ax1.barh(cba_df['Intervention'], cba_df['ROI_Percent'], color=colors)
ax1.axvline(x=100, color='gray', linestyle='--', alpha=0.5, label='Break-even (100%)')
ax1.set_xlabel('ROI (%)', fontsize=11)
ax1.set_title('Return on Investment by Intervention', fontsize=13, fontweight='bold')
ax1.legend()

for bar, val in zip(bars, cba_df['ROI_Percent']):
    ax1.text(val + 5, bar.get_y() + bar.get_height()/2, f'{val:.0f}%', va='center')
ax2 = axes[1]
scatter = ax2.scatter(cba_df['Total_5Yr_Cost_Lakhs'], cba_df['Total_5Yr_Benefit_Lakhs'],
                      s=cba_df['Units'] * 5, c=cba_df['ROI_Percent'], cmap='RdYlGn',
                      alpha=0.7, edgecolors='black')
ax2.plot([0, cba_df['Total_5Yr_Cost_Lakhs'].max()], [0, cba_df['Total_5Yr_Cost_Lakhs'].max()],
         'k--', alpha=0.3, label='Break-even line')
ax2.set_xlabel('5-Year Cost (₹ Lakhs)', fontsize=11)
ax2.set_ylabel('5-Year Benefit (₹ Lakhs)', fontsize=11)
ax2.set_title('Cost vs Benefit (Bubble = Scale)', fontsize=13, fontweight='bold')
plt.colorbar(scatter, ax=ax2, label='ROI %')
ax2.legend()

ax3 = axes[2]
cba_sorted = cba_df.sort_values('Payback_Years')
colors_payback = ['#27ae60' if pb < 2 else '#f39c12' if pb < 4 else '#e74c3c'
                  for pb in cba_sorted['Payback_Years']]
bars = ax3.barh(cba_sorted['Intervention'], cba_sorted['Payback_Years'], color=colors_payback)
ax3.axvline(x=3, color='gray', linestyle='--', alpha=0.5, label='3-Year Target')
ax3.set_xlabel('Payback Period (Years)', fontsize=11)
ax3.set_title('Investment Payback Period', fontsize=13, fontweight='bold')
ax3.legend()

plt.tight_layout()
plt.savefig('UIDAI_Analysis_Outputs/cost_benefit_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n\n" + "═"*70)
print("INVESTMENT PRIORITY RANKING:")
print("═"*70)

cba_ranked = cba_df.sort_values('ROI_Percent', ascending=False)

for rank, (_, row) in enumerate(cba_ranked.iterrows(), 1):
    status = 'HIGH PRIORITY' if row['ROI_Percent'] > 150 else 'MEDIUM' if row['ROI_Percent'] > 100 else 'LOW'
    print(f"""
   #{rank} {row['Intervention']}
      • Investment Required: ₹{row['Total_5Yr_Cost_Lakhs']:,.0f} Lakhs (5 years)
      • Expected Returns: ₹{row['Total_5Yr_Benefit_Lakhs']:,.0f} Lakhs
      • ROI: {row['ROI_Percent']:.0f}% | Payback: {row['Payback_Years']:.1f} years
      • Status: {status}
""")

print("\n" + "═"*70)
print("TOTAL INVESTMENT SUMMARY:")
print("═"*70)
print(f"""
   Total 5-Year Investment Required:  ₹{cba_df['Total_5Yr_Cost_Lakhs'].sum():,.0f} Lakhs (₹{cba_df['Total_5Yr_Cost_Lakhs'].sum()/100:.1f} Cr)
   Total 5-Year Expected Benefit:     ₹{cba_df['Total_5Yr_Benefit_Lakhs'].sum():,.0f} Lakhs (₹{cba_df['Total_5Yr_Benefit_Lakhs'].sum()/100:.1f} Cr)
   Net Benefit:                       ₹{cba_df['Net_Benefit_Lakhs'].sum():,.0f} Lakhs (₹{cba_df['Net_Benefit_Lakhs'].sum()/100:.1f} Cr)
   Overall Portfolio ROI:             {cba_df['Total_5Yr_Benefit_Lakhs'].sum()/cba_df['Total_5Yr_Cost_Lakhs'].sum()*100:.0f}%
""")

cba_df.to_csv('UIDAI_Analysis_Outputs/cost_benefit_analysis.csv', index=False)
print("Cost-Benefit Analysis saved to UIDAI_Analysis_Outputs/cost_benefit_analysis.csv")
````

### ENHANCEMENT 6.6.3: DISTRICT LEADERBOARD
Purpose: Gamified ranking of best and worst performing districts

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*15 + "DISTRICT PERFORMANCE LEADERBOARD" + " "*23 + "║")
print("╚" + "═"*78 + "╝")

print("""
   OBJECTIVE: Create a gamified ranking system for districts

   This leaderboard:
   • Ranks districts on multiple performance dimensions
   • Identifies champions for best practice replication
   • Flags underperformers for targeted support
   • Enables competitive governance benchmarking
""")

if 'AESI' in df_unified.columns:
    df_unified['Performance_Score'] = 100 - df_unified['AESI'].clip(0, 100)
else:
    df_unified['Performance_Score'] = 50

df_unified['Efficiency_Score'] = (
    df_unified['total_activity'] / (df_unified['total_activity'].max() + 1) * 100
).clip(0, 100)

if 'security_ratio' in df_unified.columns:
    df_unified['Security_Score'] = (100 - (df_unified['security_ratio'] - 1).abs() * 50).clip(0, 100)
else:
    df_unified['Security_Score'] = 50
df_unified['Leaderboard_Score'] = (
    df_unified['Performance_Score'] * 0.4 +
    df_unified['Efficiency_Score'] * 0.3 +
    df_unified['Security_Score'] * 0.3
).round(2)
df_unified['Rank'] = df_unified['Leaderboard_Score'].rank(ascending=False, method='min').astype(int)

leaderboard = df_unified[['state', 'district', 'region', 'Leaderboard_Score',
                          'Performance_Score', 'Efficiency_Score', 'Security_Score',
                          'Rank', 'total_activity']].sort_values('Rank')
def get_tier(rank, total):
    pct = rank / total * 100
    if pct <= 10: return 'Platinum'
    elif pct <= 25: return 'Gold'
    elif pct <= 50: return 'Silver'
    elif pct <= 75: return 'Bronze'
    else: return 'Needs Improvement'

leaderboard['Tier'] = leaderboard['Rank'].apply(lambda x: get_tier(x, len(leaderboard)))

print("\n" + "═"*70)
print("TOP 15 PERFORMING DISTRICTS (Champions)")
print("═"*70)

print(f"""
   {'Rank':<6} {'District':<25} {'State':<20} {'Score':>8} {'Tier':<15}
   {'─'*75}""")

for _, row in leaderboard.head(15).iterrows():
    print(f"   {row['Rank']:<6} {row['district'][:24]:<25} {row['state'][:19]:<20} {row['Leaderboard_Score']:>7.1f} {row['Tier']:<15}")

print("\n\n" + "═"*70)
print("BOTTOM 15 DISTRICTS (Need Immediate Attention)")
print("═"*70)

print(f"""
   {'Rank':<6} {'District':<25} {'State':<20} {'Score':>8} {'Gap':>8}
   {'─'*70}""")

top_score = leaderboard['Leaderboard_Score'].max()
for _, row in leaderboard.tail(15).sort_values('Rank', ascending=False).iterrows():
    gap = top_score - row['Leaderboard_Score']
    print(f"   {row['Rank']:<6} {row['district'][:24]:<25} {row['state'][:19]:<20} {row['Leaderboard_Score']:>7.1f} -{gap:>6.1f}")

print("\n\n" + "═"*70)
print("REGIONAL PERFORMANCE RANKINGS")
print("═"*70)

regional_leaderboard = leaderboard.groupby('region').agg({
    'Leaderboard_Score': 'mean',
    'Performance_Score': 'mean',
    'Efficiency_Score': 'mean',
    'Security_Score': 'mean',
    'district': 'count'
}).reset_index()

regional_leaderboard.columns = ['Region', 'Avg_Score', 'Performance', 'Efficiency', 'Security', 'Districts']
regional_leaderboard = regional_leaderboard.sort_values('Avg_Score', ascending=False)

print(f"""
   {'Rank':<6} {'Region':<15} {'Avg Score':>12} {'Districts':>12} {'Status':<20}
   {'─'*70}""")

for rank, (_, row) in enumerate(regional_leaderboard.iterrows(), 1):
    status = 'Leader' if rank == 1 else 'Lagging' if rank == len(regional_leaderboard) else 'Average'
    print(f"   {rank:<6} {row['Region']:<15} {row['Avg_Score']:>11.1f} {row['Districts']:>12} {status:<20}")

fig, axes = plt.subplots(2, 2, figsize=(16, 14))

ax1 = axes[0, 0]
top_20 = leaderboard.head(20)
colors = ['#FFD700' if '🥇' in t else '#C0C0C0' if '🥈' in t else '#CD7F32' if '🥉' in t else '#3498db'
          for t in top_20['Tier']]
ax1.barh(range(len(top_20)), top_20['Leaderboard_Score'], color=colors)
ax1.set_yticks(range(len(top_20)))
ax1.set_yticklabels([f"{d[:20]}, {s[:10]}" for d, s in zip(top_20['district'], top_20['state'])], fontsize=9)
ax1.set_xlabel('Leaderboard Score', fontsize=11)
ax1.set_title('Top 20 Performing Districts', fontsize=13, fontweight='bold')
ax1.invert_yaxis()

ax2 = axes[0, 1]
ax2.hist(leaderboard['Leaderboard_Score'], bins=30, color='#3498db', edgecolor='white', alpha=0.7)
ax2.axvline(leaderboard['Leaderboard_Score'].mean(), color='red', linestyle='--',
            label=f"Mean: {leaderboard['Leaderboard_Score'].mean():.1f}")
ax2.axvline(leaderboard['Leaderboard_Score'].median(), color='green', linestyle='--',
            label=f"Median: {leaderboard['Leaderboard_Score'].median():.1f}")
ax2.set_xlabel('Leaderboard Score', fontsize=11)
ax2.set_ylabel('Number of Districts', fontsize=11)
ax2.set_title('Score Distribution Across Districts', fontsize=13, fontweight='bold')
ax2.legend()

ax3 = axes[1, 0]
colors_region = plt.cm.Set2(np.linspace(0, 1, len(regional_leaderboard)))
ax3.barh(regional_leaderboard['Region'], regional_leaderboard['Avg_Score'], color=colors_region)
ax3.set_xlabel('Average Leaderboard Score', fontsize=11)
ax3.set_title('Regional Performance Comparison', fontsize=13, fontweight='bold')

ax4 = axes[1, 1]
tier_counts = leaderboard['Tier'].value_counts()
tier_colors = {'Platinum': '#FFD700', 'Gold': '#C0C0C0', 'Silver': '#CD7F32',
               'Bronze': '#3498db', 'Needs Improvement': '#e74c3c'}
colors_pie = [tier_colors.get(t, '#95a5a6') for t in tier_counts.index]
ax4.pie(tier_counts, labels=tier_counts.index, colors=colors_pie, autopct='%1.1f%%', startangle=90)
ax4.set_title('District Tier Distribution', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('UIDAI_Analysis_Outputs/district_leaderboard.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n\n" + "═"*70)
print("KEY INSIGHTS FROM LEADERBOARD:")
print("═"*70)
bottom_10_pct = leaderboard.tail(int(len(leaderboard) * 0.1))
avg_gap = top_score - bottom_10_pct['Leaderboard_Score'].mean()

print(f"""
   PERFORMANCE STATISTICS:
   ─────────────────────────────────────────────────────────────
   • Highest Score:     {leaderboard['Leaderboard_Score'].max():.1f} ({leaderboard.iloc[0]['district']}, {leaderboard.iloc[0]['state']})
   • Lowest Score:      {leaderboard['Leaderboard_Score'].min():.1f}
   • Mean Score:        {leaderboard['Leaderboard_Score'].mean():.1f}
   • Standard Dev:      {leaderboard['Leaderboard_Score'].std():.1f}

   TIER DISTRIBUTION:
   ─────────────────────────────────────────────────────────────""")

for tier, count in tier_counts.items():
    pct = count / len(leaderboard) * 100
    print(f"   • {tier:<25} {count:>5} districts ({pct:.1f}%)")

print(f"""

   RECOMMENDATIONS:
   ─────────────────────────────────────────────────────────────
   • Study top 10% (Platinum) districts for best practices
   • Focus improvement efforts on bottom 25% districts
   • Average score gap between top and bottom: {avg_gap:.1f} points
   • Regional champion: {regional_leaderboard.iloc[0]['Region']} (Avg: {regional_leaderboard.iloc[0]['Avg_Score']:.1f})
""")

leaderboard.to_csv('UIDAI_Analysis_Outputs/district_leaderboard.csv', index=False)
print("District Leaderboard saved to UIDAI_Analysis_Outputs/district_leaderboard.csv")
````

### ENHANCEMENT 6.6.4:  WHAT-IF SCENARIO SIMULATOR
Purpose: Interactive "What if we add 100 centers to Northeast?" projections

````python
print("╔" + "═"*78 + "╗")
print("║" + " "*12 + "WHAT-IF SCENARIO SIMULATOR" + " "*30 + "║")
print("╚" + "═"*78 + "╝")
print("""
   OBJECTIVE: Project impact of policy interventions

   This simulator answers questions like:
   • "What if we add 100 centers to Northeast?"
   • "What if we increase capacity by 20% in Bihar?"
   • "What if we deploy mobile units to underserved areas?"
""")

regional_baseline = df_unified.groupby('region').agg({
    'total_enrollment': 'sum',
    'total_activity': 'sum',
    'AESI': 'mean',
    'district': 'count',
    'total_demo': 'sum',
    'total_bio': 'sum'
}).reset_index()

regional_baseline.columns = ['Region', 'Current_Enrollment', 'Current_Activity',
                            'Current_Risk_Score', 'Districts', 'Demo_Updates', 'Bio_Updates']
regional_baseline['Current_Capacity'] = regional_baseline['Current_Enrollment'] / regional_baseline['Districts']
regional_baseline['Current_Coverage_Index'] = (regional_baseline['Current_Activity'] /
                                                regional_baseline['Current_Activity'].sum() * 100)
print("\n" + "═"*70)
print("CURRENT REGIONAL BASELINE:")
print("═"*70)

print(f"""
   {'Region':<15} {'Enrollment':>15} {'Activity':>15} {'Risk Score':>12} {'Capacity/District':>18}
   {'─'*75}""")

for _, row in regional_baseline.sort_values('Current_Activity', ascending=False).iterrows():
    print(f"   {row['Region']:<15} {row['Current_Enrollment']:>14,.0f} {row['Current_Activity']:>14,.0f} {row['Current_Risk_Score']:>11.1f} {row['Current_Capacity']:>17,.0f}")

def simulate_scenario(scenario_name, region_changes):
    results = regional_baseline.copy()

    ENROLLMENT_PER_NEW_CENTER = 5000
    ACTIVITY_PER_CENTER = 8000
    RISK_REDUCTION_PER_CENTER = 0.5

    for region, changes in region_changes.items():
        idx = results[results['Region'] == region].index
        if len(idx) == 0:
            continue
        idx = idx[0]

        new_centers = changes.get('centers', 0)
        capacity_increase = changes.get('capacity_increase', 0) / 100

        results.loc[idx, 'New_Centers'] = new_centers
        results.loc[idx, 'Projected_Enrollment'] = (
            results.loc[idx, 'Current_Enrollment'] * (1 + capacity_increase) +
            new_centers * ENROLLMENT_PER_NEW_CENTER
        )
        results.loc[idx, 'Projected_Activity'] = (
            results.loc[idx, 'Current_Activity'] * (1 + capacity_increase) +
            new_centers * ACTIVITY_PER_CENTER
        )
        results.loc[idx, 'Projected_Risk'] = max(
            10,
            results.loc[idx, 'Current_Risk_Score'] - (new_centers * RISK_REDUCTION_PER_CENTER)
        )
    for col in ['Projected_Enrollment', 'Projected_Activity', 'Projected_Risk', 'New_Centers']:
        if col not in results.columns:
            results[col] = 0
        results[col] = results[col].fillna(
            results['Current_Enrollment'] if 'Enrollment' in col else
            results['Current_Activity'] if 'Activity' in col else
            results['Current_Risk_Score'] if 'Risk' in col else 0
        )
    results['Enrollment_Change'] = results['Projected_Enrollment'] - results['Current_Enrollment']
    results['Activity_Change'] = results['Projected_Activity'] - results['Current_Activity']
    results['Risk_Change'] = results['Projected_Risk'] - results['Current_Risk_Score']

    return results

scenarios = {
    'Scenario 1: Northeast Expansion': {
        'Northeast': {'centers': 100, 'capacity_increase': 25}
    },
    'Scenario 2: Bihar-Jharkhand Focus': {
        'East': {'centers': 150, 'capacity_increase': 30}
    },
    'Scenario 3: Pan-India Coverage Push': {
        'Northeast': {'centers': 50, 'capacity_increase': 15},
        'East': {'centers': 75, 'capacity_increase': 15},
        'Central': {'centers': 50, 'capacity_increase': 10},
        'North': {'centers': 100, 'capacity_increase': 10}
    },
    'Scenario 4: High-Risk Region Priority': {
        'Northeast': {'centers': 75, 'capacity_increase': 20},
        'Central': {'centers': 75, 'capacity_increase': 20}
    }
}

scenario_results = {}

print("\n\n" + "═"*70)
print("SCENARIO SIMULATION RESULTS:")
print("═"*70)

for scenario_name, changes in scenarios.items():
    result = simulate_scenario(scenario_name, changes)
    scenario_results[scenario_name] = result

    total_new_enrollment = result['Enrollment_Change'].sum()
    total_new_activity = result['Activity_Change'].sum()
    avg_risk_reduction = result['Risk_Change'].mean()
    total_new_centers = result['New_Centers'].sum() if 'New_Centers' in result.columns else 0

    print(f"""
   ┌─────────────────────────────────────────────────────────────────────┐
   │ {scenario_name:<67}                                                 │
   ├─────────────────────────────────────────────────────────────────────┤
   │ New Centers:           {total_new_centers:>8,.0f}                   │
   │ Additional Enrollments: {total_new_enrollment:>12,.0f}              │
   │ Additional Activity:    {total_new_activity:>12,.0f}                │
   │ Avg Risk Reduction:    {avg_risk_reduction:>8.1f} points            │
   └─────────────────────────────────────────────────────────────────────┘""")

fig, axes = plt.subplots(2, 2, figsize=(16, 14))

ax1 = axes[0, 0]
scenario_names = list(scenarios.keys())
enrollment_impacts = [scenario_results[s]['Enrollment_Change'].sum() / 1000 for s in scenario_names]
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(scenario_names)))
bars = ax1.bar(range(len(scenario_names)), enrollment_impacts, color=colors)
ax1.set_xticks(range(len(scenario_names)))
ax1.set_xticklabels([s.replace('Scenario ', 'S').split(':')[0] for s in scenario_names], fontsize=10)
ax1.set_ylabel('Additional Enrollments (Thousands)', fontsize=11)
ax1.set_title('Projected Enrollment Impact by Scenario', fontsize=13, fontweight='bold')

for bar, val in zip(bars, enrollment_impacts):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10, f'+{val:.0f}K',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

ax2 = axes[0, 1]
risk_impacts = [scenario_results[s]['Risk_Change'].mean() for s in scenario_names]
colors_risk = ['#27ae60' if r < 0 else '#e74c3c' for r in risk_impacts]
bars = ax2.bar(range(len(scenario_names)), risk_impacts, color=colors_risk)
ax2.set_xticks(range(len(scenario_names)))
ax2.set_xticklabels([s.replace('Scenario ', 'S').split(':')[0] for s in scenario_names], fontsize=10)
ax2.set_ylabel('Average Risk Score Change', fontsize=11)
ax2.set_title('Projected Risk Reduction by Scenario', fontsize=13, fontweight='bold')
ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.5)

ax3 = axes[1, 0]
best_scenario = max(scenarios.keys(), key=lambda s: scenario_results[s]['Enrollment_Change'].sum())
best_result = scenario_results[best_scenario]

x = np.arange(len(best_result))
width = 0.35
bars1 = ax3.bar(x - width/2, best_result['Current_Enrollment']/1000, width, label='Current', color='#3498db')
bars2 = ax3.bar(x + width/2, best_result['Projected_Enrollment']/1000, width, label='Projected', color='#27ae60')
ax3.set_xticks(x)
ax3.set_xticklabels(best_result['Region'], fontsize=9, rotation=45, ha='right')
ax3.set_ylabel('Enrollment (Thousands)', fontsize=11)
ax3.set_title(f'Current vs Projected: {best_scenario.split(":")[1].strip()}', fontsize=12, fontweight='bold')
ax3.legend()

ax4 = axes[1, 1]
investments = []
impacts = []
for s in scenario_names:
    result = scenario_results[s]
    total_centers = result['New_Centers'].sum() if 'New_Centers' in result.columns else 0
    investment = total_centers * 15
    impact = result['Enrollment_Change'].sum() / 1000
    investments.append(investment)
    impacts.append(impact)

scatter = ax4.scatter(investments, impacts, s=200, c=range(len(scenario_names)),
                      cmap='viridis', alpha=0.7, edgecolors='black')
for i, name in enumerate(scenario_names):
    ax4.annotate(name.split(':')[0], (investments[i], impacts[i]),
                 xytext=(10, 5), textcoords='offset points', fontsize=9)

ax4.set_xlabel('Investment (₹ Lakhs)', fontsize=11)
ax4.set_ylabel('Impact (Thousands of Enrollments)', fontsize=11)
ax4.set_title('Investment vs Impact Matrix', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('UIDAI_Analysis_Outputs/scenario_simulator.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n\n" + "═"*70)
print("DETAILED SCENARIO: NORTHEAST EXPANSION (100 Centers)")
print("═"*70)

ne_result = scenario_results['Scenario 1: Northeast Expansion']

print(f"""
   NORTHEAST EXPANSION IMPACT ANALYSIS:
   ─────────────────────────────────────────────────────────────

   Current State:
   • Enrollment:     {ne_result[ne_result['Region'] == 'Northeast']['Current_Enrollment'].values[0]:>12,.0f}
   • Activity:       {ne_result[ne_result['Region'] == 'Northeast']['Current_Activity'].values[0]:>12,.0f}
   • Risk Score:     {ne_result[ne_result['Region'] == 'Northeast']['Current_Risk_Score'].values[0]:>12.1f}

   After Adding 100 Centers (+25% Capacity):
   • Enrollment:     {ne_result[ne_result['Region'] == 'Northeast']['Projected_Enrollment'].values[0]:>12,.0f}
   • Activity:       {ne_result[ne_result['Region'] == 'Northeast']['Projected_Activity'].values[0]:>12,.0f}
   • Risk Score:     {ne_result[ne_result['Region'] == 'Northeast']['Projected_Risk'].values[0]:>12.1f}

   Net Impact:
   • +{ne_result[ne_result['Region'] == 'Northeast']['Enrollment_Change'].values[0]:,.0f} new enrollments
   • +{ne_result[ne_result['Region'] == 'Northeast']['Activity_Change'].values[0]:,.0f} additional activities
   • {ne_result[ne_result['Region'] == 'Northeast']['Risk_Change'].values[0]:.1f} point risk reduction

   Investment Required: ~₹{100 * 15:,} Lakhs (₹{100 * 15 / 100:.1f} Cr)
   Cost per Enrollment: ₹{100 * 15 * 100000 / ne_result[ne_result['Region'] == 'Northeast']['Enrollment_Change'].values[0]:.0f}
""")


print("\n" + "═"*70)
print("SCENARIO-BASED RECOMMENDATIONS:")
print("═"*70)

scenario_efficiency = {}
for s in scenario_names:
    result = scenario_results[s]
    total_centers = result['New_Centers'].sum() if 'New_Centers' in result.columns else 1
    investment = total_centers * 15
    impact = result['Enrollment_Change'].sum()
    efficiency = impact / (investment + 1)
    scenario_efficiency[s] = {
        'investment': investment,
        'impact': impact,
        'efficiency': efficiency,
        'risk_reduction': result['Risk_Change'].mean()
    }

best_efficiency = max(scenario_efficiency.items(), key=lambda x: x[1]['efficiency'])
best_impact = max(scenario_efficiency.items(), key=lambda x: x[1]['impact'])
best_risk = min(scenario_efficiency.items(), key=lambda x: x[1]['risk_reduction'])

print(f"""
   BEST FOR EFFICIENCY (Impact per ₹ Invested):
      {best_efficiency[0]}
      • Efficiency Score: {best_efficiency[1]['efficiency']:.1f}

   BEST FOR MAXIMUM IMPACT:
      {best_impact[0]}
      • Total New Enrollments: {best_impact[1]['impact']:,.0f}

   BEST FOR RISK REDUCTION:
      {best_risk[0]}
      • Avg Risk Reduction: {best_risk[1]['risk_reduction']:.1f} points

   RECOMMENDED STRATEGY:
      For limited budget: {best_efficiency[0].split(':')[1].strip()}
      For maximum coverage: {best_impact[0].split(':')[1].strip()}
      For security priority: {best_risk[0].split(':')[1].strip()}
""")
for name, result in scenario_results.items():
    safe_name = name.replace(':', '_').replace(' ', '_').lower()
    result.to_csv(f'UIDAI_Analysis_Outputs/scenario_{safe_name[:30]}.csv', index=False)

print("\n What-If Scenario Simulations Complete!")
print("   Scenario results saved to UIDAI_Analysis_Outputs/")
````

## SECTION 7.3: Urgency Matrix for Proactive Aadhaar Governance

> The Urgency Matrix converts analytical insights into a tiered, action-oriented governance framework. Districts are classified into three urgency tiers using the **Aadhaar Enrollment Stress Index (AESI)** and supporting capacity indicators, enabling proportional and timely administrative response.

### Urgency Matrix Structure

| Tier | Classification Criteria | Response Timeline | Governance Focus | Coverage Scope |
|------|------------------------|-------------------|------------------|----------------|
|  **Tier 1 – Immediate** | AESI > 75 OR High Risk – Low Capacity | 0–30 days | Risk validation, supervisory oversight, rapid capacity support | Highest-risk districts |
|  **Tier 2 – Urgent** | AESI 50–75 OR rising stress trends | 30–90 days | Training reinforcement, compliance strengthening, infrastructure readiness | Emerging-risk districts |
|  **Tier 3 – Planned** | AESI 25–50 OR capacity gaps | 90–180 days | Expansion planning, mobile outreach, awareness programs | Underserved regions |

---

###  Tier 1: Immediate Intervention (0–30 Days)

**Objective:** Stabilize operations and validate risk signals in high-stress regions.

**Indicative Actions:**
- Prioritized audits of high-risk enrollment centers
- Temporary supervisory oversight and monitoring
- Deployment of additional biometric and technical support resources

> *Note: Actions are indicative and designed to support evidence-based administrative decision-making.*

---

###  Tier 2: Preventive Reinforcement (30–90 Days)

**Objective:** Prevent escalation of operational stress through targeted capacity strengthening.

**Indicative Actions:**
- Refresher training for enrollment operators
- Infrastructure readiness assessments
- Enhanced compliance monitoring

---

###  Tier 3: Planned Capacity Expansion (90–180 Days)

**Objective:** Improve long-term coverage and service resilience in under-served regions.

**Indicative Actions:**
- Mobile enrollment and update camps
- Targeted awareness initiatives
- Strategic capacity expansion planning

### Section 7.3b: Urgency Matrix – Operationalization Logic

The Urgency Matrix operationalizes the Aadhaar Enrollment Stress Index (AESI) into tiered administrative urgency levels. The classification is **indicative**, designed to support prioritization and sequencing of governance response rather than prescribe enforcement actions.

Tier thresholds are derived from relative AESI distributions to ensure scalable, non-arbitrary decision support.
